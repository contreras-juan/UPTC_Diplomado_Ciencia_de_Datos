El diccionario de datos, tambin conocido como "Data Dictionary" en ingls, es una herramienta fundamental en el campo de la Ciencia de Datos y el Anlisis Exploratorio de Datos (EDA). Se trata de un documento que proporciona informacin detallada y descriptiva sobre los datos utilizados en un proyecto o conjunto de datos. El diccionario de datos es una herramienta esencial que contribuye significativamente a la transparencia, calidad y eficiencia en el manejo y anlisis de datos en proyectos.

* **Nombres de variables o columnas:** Enumera todas las variables o campos presentes en el conjunto de datos.

* **Tipo de datos:** Indica el tipo de informacin que contiene cada variable, como numrico, categrico, fecha, etc.

* **Descripcin de variables:** Proporciona una explicacin detallada de lo que representa cada variable. Esto puede incluir su significado, unidades de medida y cualquier otra informacin relevante.

* **Rangos y valores permitidos:** Especifica los rangos de valores que puede tener cada variable y cualquier restriccin o regla asociada.

* **Origen de datos:** Indica la fuente de donde provienen los datos, cmo se recopilaron y cualquier proceso de transformacin aplicado.

* **Formato de representacin:** Define cmo se almacena y muestra la informacin, como el formato de fecha, el nmero de decimales, etc.

La importancia de conocer el diccionario de datos radica en varios aspectos:

1. **Comprensin del contexto:** Facilita la comprensin del contexto y el significado de cada variable, permitiendo a los analistas y cientficos de datos interpretar adecuadamente los resultados.

2. **Consistencia y calidad de datos:** Ayuda a garantizar la consistencia y calidad de los datos al proporcionar informacin sobre los formatos y rangos permitidos, lo que puede ser crucial para identificar posibles problemas o anomalas en los datos.

3. **Colaboracin:** Facilita la colaboracin entre diferentes equipos o personas involucradas en un proyecto al proporcionar una referencia comn y comprensible sobre la estructura y contenido de los datos.

4. **Documentacin:** Sirve como una valiosa documentacin para futuros anlisis y proyectos, permitiendo a otros entender rpidamente la naturaleza de los datos sin tener que explorar el conjunto de datos de forma exhaustiva.

### 2.1. Estadstica descriptiva bsica

---
Estadsticos de tendencia central

---


Los estadsticos de tendencia central son medidas que representan el valor central o tpico de un conjunto de datos. Son tiles para resumir la distribucin de los datos y proporcionar una idea general de dnde se concentran los datos. Los tres estadsticos de tendencia central ms comunes son la media, la mediana y la moda.

1. **Media o promedio**: La media es el valor obtenido al sumar todos los valores de un conjunto de datos y luego dividir esa suma por el nmero total de valores. Se calcula con la siguiente ecuacin:

$$\mu = \frac{\sum_{i=1}^{n} x_i}{n}$$

2. **Mediana**: La mediana es el valor que separa el conjunto de datos ordenados en dos partes iguales: la mitad de los datos estn por encima de la mediana y la otra mitad estn por debajo. Si el nmero de datos es impar, la mediana es simplemente el valor en el centro del conjunto de datos. Si el nmero de datos es par, la mediana es el promedio de los dos valores centrales.

$$
\text{Mediana} = \begin{cases}
x_{\frac{n+1}{2}} & \text{si } n \text{ es impar} \\
\frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2} & \text{si } n \text{ es par}
\end{cases}
$$

3. **Moda**: La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Puede haber ms de una moda (en cuyo caso los datos son multimodales) o puede no haber ninguna moda (cuando todos los valores tienen la misma frecuencia).

$$
\text{Moda} = \arg \max_{x} f(x)
$$


---
Medidas de dispersin

---
Las medidas de dispersin son estadsticas que proporcionan informacin sobre la variabilidad o dispersin de un conjunto de datos. Son tiles para comprender la distribucin de los datos y la distancia entre los valores individuales y el centro del conjunto de datos. Las principales medidas de dispersin incluyen la varianza, la desviacin estndar, el rango y el rango intercuartlico.

1. **Varianza**: La varianza mide la dispersin de los datos alrededor de la media. Es la media de los cuadrados de las desviaciones de cada valor respecto a la media del conjunto de datos.

$$
\sigma^2 = \frac{\sum_{i=1}^{n} (x_i - \mu)^2}{n}
$$

2. **Desviacin estndar**: La desviacin estndar es simplemente la raz cuadrada de la varianza. Proporciona una medida de dispersin en las mismas unidades que los datos originales, lo que la hace ms interpretable.

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
$$

3. **Rango**:El rango es la diferencia entre el valor mximo y el valor mnimo en un conjunto de datos. Es una medida simple de dispersin pero puede ser afectada por valores atpicos.

$$
\text{Rango} = x_{\text{mx}} - x_{\text{mn}}
$$

4. **Rango Intercuartlico (IQR)**: El rango intercuartlico es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Es una medida de dispersin ms robusta que el rango, ya que no se ve afectada por valores extremos.

5. **Cuartiles**: Los cuartiles son medidas estadsticas que dividen un conjunto de datos ordenados en cuatro partes iguales. Se utilizan para describir la distribucin de los datos y proporcionar informacin sobre la dispersin y la posicin de los valores dentro del conjunto de datos. Los tres cuartiles principales son:

$$
\text{RIC} = Q3 - Q1
$$

$$
Q1 = x_{\frac{n+1}{4}}
$$

$$
Q2 = \begin{cases}
x_{\frac{n+1}{2}} & \text{si } n \text{ es impar} \\
\frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2} & \text{si } n \text{ es par}
\end{cases}
$$

$$
Q3 = x_{\frac{3(n+1)}{4}}
$$

---
Coeficiente de variacion (CV)

---

El coeficiente de variacin (CV) es una medida estadstica que se utiliza para cuantificar la variabilidad relativa de una variable en relacin con su media. Se calcula como la desviacin estndar de la variable dividida por su media, y se expresa como un porcentaje. La frmula para el coeficiente de variacin (CV) es la siguiente:

$$ CV (X) = \frac{std(X)*100}{mean(X)} $$

* $CV (X)$ es el coeficiente de variacin de la variable $X$
* $std(X)$ es la desviacin estndar de la variable $X$
* $mean(X)$ es la media de la variable $X$

El coeficiente de variacin es til cuando se desea comparar la variabilidad de diferentes conjuntos de datos que pueden tener unidades de medida diferentes o escalas distintas. Por ejemplo, si se estn comparando dos distribuciones de ingresos en diferentes pases, el CV permite evaluar cul de las distribuciones es ms variable en relacin con su media, independientemente de las unidades de moneda utilizadas en cada pas.

Un coeficiente de variacin bajo indica que la variabilidad de la variable es relativamente pequea en relacin con su media, mientras que un coeficiente de variacin alto indica una mayor variabilidad relativa en comparacin con la media. El CV se expresa como un porcentaje para facilitar su interpretacin y comparacin entre diferentes variables o conjuntos de datos.


El rango intercuartlico (IQR, por sus siglas en ingls) y los cuartiles son medidas estadsticas que se utilizan para comprender la distribucin de un conjunto de datos y para identificar la variabilidad en los datos.

* Los cuartiles son tiles para entender cmo se distribuyen los datos y proporcionan informacin sobre la dispersin y la forma de la distribucin.

* El IQR es una medida de dispersin que se centra en la mitad central de los datos. Es menos sensible a valores atpicos que el rango completo y proporciona una indicacin de la variabilidad de los datos en el intervalo intercuartlico. Es indispensable para identificar datos como outliers.

```python
# Calculo de los cuantiles
cuantiles = df_house.quantile(q=[0.75, 0.50, 0.25], numeric_only=True)
cuantiles = cuantiles.transpose().rename_axis('Variable').reset_index()
cuantiles
```

```python
# Clculo del IQR
cuantiles['IQR'] = cuantiles[0.75] - cuantiles[0.25]

# Clculo del limite inferior y superior
cuantiles['Limite inferior'] = cuantiles[0.25] - 1.5*cuantiles['IQR']
cuantiles['Limite superior'] = cuantiles[0.75] + 1.5*cuantiles['IQR']
cuantiles
```

Calculemos ahora, el coeficiente de variacin de cada una de nuestras variables.

```python
# Fincin para calcular la variabilidad de nuestro dataset numrico
def CV_df(df):
    categorical = df.select_dtypes(include='category').columns
    df_numerico = df.drop(columns=categorical)
    non_categorical = df_numerico.columns

    CV = {columna: (df_numerico[columna].std() * 100) / df_numerico[columna].mean() for columna in non_categorical}
    CV = pd.DataFrame([CV], columns=non_categorical)
    return CV

CV_df(df_house)
```

Anlisis Bivariado de correlacin (Scipy)

---

SciPy es una biblioteca de Python utilizada para realizar clculos cientficos y matemticos. Ofrece una amplia gama de funciones y herramientas para el anlisis de datos, la optimizacin, la interpolacin, entre otros. En el contexto del anlisis de datos, SciPy proporciona mtodos estadsticos clave que permiten evaluar la relacin entre variables y determinar su significancia.

Cuando se analizan datos y se buscan relaciones entre variables, es crucial respaldar las conclusiones con estadstica inferencial. Una medida comn para evaluar la relacin entre variables es el coeficiente de correlacin, que puede calcularse utilizando mtodos disponibles en SciPy como:
* `stats.pearsonr`: Coeficiente de correlacin $r$ de Pearson.
* `stats.spearmanr`: Coeficiente de correlacin $\rho$ de Spearman.
* `stats.kendalltau`: Coeficiente de correlacin $\tau$ de Kendall.

Estas funciones devuelven no solo el coeficiente de correlacin, sino tambin el valor $p$ asociado. Si el **p-valor $\lt \alpha$** (nivel de significancia), esto implica que la correlacin entre variables es estadsticamente significativa y que el valor de los estadsticos de correlacin dan cuenta de la colinealidad de las variables. De lo contrario, la correlacin puede deberse a la aleatoriedad de los datos independientemente de su valor.


```python
from scipy import stats

columns = df_numerico.columns
l = len(columns)
corr_matrix = [ [0 for _ in range(l)] for _ in range(l) ]
p_matrix    = [ [0 for _ in range(l)] for _ in range(l) ]

for i, columna_1 in enumerate(columns):
  for j, columna_2 in enumerate(columns):
    corr_matrix[i][j], p_matrix[i][j] = stats.pearsonr(df_numerico[columna_1],df_numerico[columna_2])
    corr_matrix[i][j] = round(corr_matrix[i][j],2)
    p_matrix[i][j]    = round(p_matrix[i][j],2)

print(columns)

fig, ax = plt.subplots(1,2,figsize = (14,6))
sns.heatmap(data = corr_matrix,center=0, annot = True, ax=ax[0])
sns.heatmap(data = p_matrix,   center=0, annot = True, ax=ax[1])
plt.show()
```

Anlisis por agrupacin segn categoras

Hacer agrupaciones de datos en un Anlisis Exploratorio de Datos (EDA, por sus siglas en ingls) con Pandas es una tcnica fundamental para comprender la estructura y las caractersticas de un conjunto de datos. Las agrupaciones permiten analizar y resumir la informacin de manera ms significativa al agrupar los datos segn una o ms variables y luego aplicar funciones de resumen o anlisis a cada grupo. Dentro de las utilidades de generar agrupaciones en los datos encontramos:

* Resumen de datos por categoras.
* Anlisis comparativo.
* Identificacin de patrones y tendencias.
* Segmentacin de datos.
* Preparacin de datos para visualizacin o modelado.

El producto de una agrupacin en pandas en un objeto groupby, el cual es un iterable que contiene el nombre de cada grupo y adems un subconjunto o dataframe para cada grupo.

```python
grupos = df_house.groupby('bedrooms')
```

```python
for nombre, grupo in grupos:
  print('Nombre del grupo:', nombre)
  print(grupo)
```

Para obtener un grupo en especfico, empleamos el mtodo `get_group()`.

```python
grupos_ = []
for i in range(len(grupos)):
  grupo = grupos.get_group(i+1)
  grupos_.append(grupo)
```

```python
grupos_[0]
```

Ya que cada agrupacin corresponde a un dataframe, nos es posible ejecutar sobre este todas las operacines que deseemos. Por ejemplo, veamos el promedio de una agrupacin.

```python
# Estadsticos clsicos para una agrupacin completa
grupos.mean()
```

```python
# Estadstica descriptiva bsica para un grupo en concreto
grupos_[0].describe()
```

* En promedio, qu nmero de habitaciones posee el precio por rea mayor?

```python
bed_promedios = df_house.groupby('bedrooms').mean()['price_per_area']
jerarquia = bed_promedios.sort_values(ascending=False)

# Trazar el grfico de barras
plt.figure(figsize=(6, 4))
jerarquia.plot(kind='bar')
plt.title('Precio promedio por rea segn la cantidad de dormitorios')
plt.xlabel('Cantidad de dormitorios')
plt.ylabel('Precio promedio por rea')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
bed_promedios = df_house.groupby('bedrooms').std()['price_per_area']
jerarquia = bed_promedios.sort_values(ascending=False)

# Trazar el grfico de barras
plt.figure(figsize=(6, 4))
jerarquia.plot(kind='bar')
plt.title('STD. del Precio por rea segn la cantidad de dormitorios')
plt.xlabel('Cantidad de dormitorios')
plt.ylabel('Desviacin estndard del PPA ')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Para complementar los anlisis, es til realizar agrupaciones dentro de las agrupaciones. Para ello, es posible emplear el mtodo .agg(). Por ejemplo, realicemos una consulta que responda la siguiente pregunta:

* Cul es el menor precio y la menor rea para cada nmero de habitaciones?

```python
# Agregacin mltiple
grupos = df_house.groupby('bedrooms')
variables_de_interes = {'price': 'mean', 'area': 'min'}
sub_grupos = grupos.agg(variables_de_interes)

sub_grupos
```

En Pandas, tambin es posible realizar consultas anidadas. Intentemos contestar la siguiente pregunta:
* Cul es el mnimo y mximo del rea en conjunto con el promedio y la mediana por nmero de habitaciones?

```python
# Agregacin anidada
consulta = {'area' : ['min' , 'max'], 'price': ['mean','median']}
consulta_anidada = grupos.agg(consulta)

consulta_anidada
```

Para finalizar, recordemos los siguientes mtodos:

* `df[ [lista de variables] ]` : Filtra el dataframe de acuerdo a un grupo de variables.

* `df.sort_values(by = variable, ascending = True/False)` : Organiza el dataframe deacuerdo a los valores de una variable en forma ascendente o descendente. Por defecto, los ordena de manera ascendente.

* `df[inicio, final, paso]` : Selecciona aquellos regstros del dataframe a partir del regstro `inicio` hasta `final` con condicin de paso `paso`.

* `df.loc[condiciones]` : Permite seleccionar regstros del dataframe a partir de un conjunto de condiciones tales como nombres de variables, ndices, condiciones de fltro o combinaciones de estas.

Teniendo en cuenta estos mtodos, busquemos el inmueble que cumple las siguientes caractersticas:
