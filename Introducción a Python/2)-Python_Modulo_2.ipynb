### 1.1. Definicin de funciones

---
Definicin de funciones

---

Las funciones en Python son bloques de cdigo reutilizables que realizan una tarea especfica. Su uso es fundamental para organizar y modularizar el cdigo, facilitando la lectura, mantenimiento y reutilizacin.

La sintaxis bsica de una funcin es la siguiente, donde para la definicin de una funcin se usa el comando `def`.

```python
# Sintaxis bsica de una funcion

def nombre_de_mi_funcion(parametros):
  # Bloque de cdigo a usar como funcin (reutilizable)
```

Las funciones que no requieren el ingreso de parmetros por parte del programador, son ejecutadas mediante su nombre y un par cerrado de parntesis en la forma `nombre_de_mi_funcion()`

```python
# Ejempo bsico de una funcin

def imprimir_mensaje():
  print('Mensaje por defecto')

imprimir_mensaje()
```

```python
# Ejemplo simple de una funcin con parmetros

def saludar(nombre):
    """Esta funcin saluda a la persona proporcionada."""
    print(f"Hola, {nombre}!")

# Uso de la funcin
saludar("Alice")
```

```python
# Ejemplo simple de una funcin con parmetros

def operacion(a,b,c):
    variable_interna = (a + b)^c + a**3
    print("Resultado = ", variable_interna )

# Uso de la funcin
operacion(3,4,2)
```

Las funcines cuando son ejecutadas, corren el cdigo que tienen dentro de s, no obstante estos bloques de cdigo pueden o no retornar un valor producto de la ejecucin del bloque. Para que una funcin retorne un resultado producto de la ejecucin de una secuencia de procedimientos es necesario especificarlo mediante la palabra reservada `return`.

```python
# Ejemplo de funcin que retorna un valor

def operacion_no_return(a,b,c):
    variable_interna = (a + b)^c + a**3

def operacion_return(a,b,c):
    variable_interna = (a + b)^c + a**3
    return variable_interna

# Uso de la funcin
nueva_variable = operacion_return(3,4,2)

print("el resultado de la operacion es = ", nueva_variable)
print("el resultado de la operacion es = ", operacion_no_return(3,4,2))
```

Las funcines, como bloques de procedimientos, pueden ser definidas para operar con valores por defecto, los cuales son valores que tomarn los parmetros requerdos por la funcin en caso de que el programador no estipule explcitamente el valor de stos.

```python
# Funcin con valores por defecto

def suma(a, b=0):
    return a + b

# Uso de la funcin con parmetro opcional
resultado1 = suma(3, 5)
resultado2 = suma(3)

print(f"Suma con dos nmeros: {resultado1}")
print(f"Suma con un nmero (b asume valor predeterminado): {resultado2}")
```

Las funciones pueden tomar como argumentos, o parmetros, cualquier tipo de dato siempre y cuando este valor de entrada coincda con los requerimientos de tipo de las operaciones efectuadas dentro de la funcin.

```python
# Funcin con ciclo for()

def suma_acumulativa(min, max):
  sum = 0
  sust = 0
  proceso_1 = []
  proceso_2 = []

  for x in range(min, max):
    sum += x
    sust -=x

    proceso_1.append(sum)
    proceso_2.append(sust)

  return sum, proceso_1, proceso_2

result = suma_acumulativa(1, 10)
suma, progreso_1, progreso_2 = suma_acumulativa(1, 10)

print(result)
print(suma)
print(progreso_1)
print(progreso_2)
```

```python
# Funcin con una lista y un nmero como parmetros

def operacion_con_lista(lista, multiplicador):
    """
    Esta funcin toma una lista y un entero como argumentos,
    multiplica cada elemento de la lista por el multiplicador
    y devuelve una nueva lista con los resultados.
    """
    resultado = []

    for elemento in lista:
        resultado.append(elemento * multiplicador)

    return resultado

# Ejemplo de uso
mi_lista = [1, 2, 3, 4, 5]
resultado_operacion = operacion_con_lista(mi_lista, 3)

print(f"Lista original: {mi_lista}")
print(f"Resultado de la operacin: {resultado_operacion}")
```

```python
# Funcin con una lista, un diccionario y un nmero como parmetros

def operacion_con_lista_diccionario(lista, diccionario, numero):
    """
    Esta funcin toma una lista, un diccionario y un nmero como argumentos,
    realiza una operacin combinada con estos datos,
    y devuelve el resultado.
    """
    resultado = []

    for elemento in lista:
        resultado.append(elemento * numero)

    for key, value in diccionario.items():
        resultado.append(value - numero)

    return resultado

# Ejemplo de uso
mi_lista = [1, 2, 3, 4, 5]
mi_diccionario = {'a': 10, 'b': 20, 'c': 30}
mi_numero = 5

resultado_operacion = operacion_con_lista_diccionario(mi_lista, mi_diccionario, mi_numero)

print(f"Lista original: {mi_lista}")
print(f"Diccionario original: {mi_diccionario}")
print(f"Nmero original: {mi_numero}")
print(f"Resultado de la operacin: {resultado_operacion}")
```

### 1.2. Scope y funciones

---
Scope y funciones

---

El scope (alcance) en Python se refiere a la regin de un programa donde una variable es vlida y puede ser accedida. Establece las reglas para la visibilidad y accesibilidad de las variables en diferentes partes del cdigo. El scope ayuda a organizar y controlar el flujo de datos en un programa al definir dnde una variable puede ser creada, modificada o utilizada.

* **Local Scope  (mbito Local)**: Las variables definidas dentro de una funcin tienen un alcance local y solo son accesibles desde esa funcin. Estas variables existen solo mientras se ejecuta la funcin y desaparecen cuando la funcin termina su ejecucin.

* **Enclosed Scope (mbito Encerrado)**: Tambin conocido como "closure", este alcance se aplica cuando una funcin est anidada dentro de otra. Las variables en el mbito exterior (encerrado) de la funcin interna pueden ser accedidas y, en algunos casos, modificadas utilizando la palabra clave `nonlocal`.

* **Global Scope (mbito Global)**: Las variables definidas fuera de todas las funciones tienen un alcance global y son accesibles desde cualquier parte del cdigo, incluyendo funciones. Se pueden modificar dentro de funciones utilizando la palabra clave global.

```python
# Scope local

def funcion_local(a , b):
    variable_local = 10
    resultado = variable_local * (a + b) ** 3
    print(f"Variable local dentro de la funcin: {variable_local}")

funcion_local(13, 8)
# print(f"Variable local fuera de la funcin: {variable_local}")  # Error: variable_local no definida fuera de la funcin
```

```python
# Scope Closure

def funcion_exterior():
    variable_exterior = 20

    def funcion_interior():
        nonlocal variable_exterior
        variable_exterior += 5
        print(f"Variable exterior dentro de la funcin interna: {variable_exterior}")

    funcion_interior()
    print(f"Variable exterior fuera de la funcin interna: {variable_exterior}")

funcion_exterior()
```

```python
# Scope global

iteraciones = 10

def funcion(a,b):

  resultado = 0

  for i in range(iteraciones):
    resultado += (a*b) + 9/a

  return resultado

funcion(1,2)
```

En Python, la palabra clave `global` se utiliza para indicar que una variable que se est utilizando dentro de una funcin se refiere a la misma variable definida en el alcance global (fuera de la funcin), en lugar de crear una nueva variable local con el mismo nombre.

Cuando se declara una variable como global dentro de una funcin, se le permite a esa funcin modificar el valor de la variable global. Sin la palabra clave global, si se intenta modificar una variable dentro de una funcin, Python crear una nueva variable local con ese nombre en lugar de modificar la variable global.

```python
# Scope global modificado

variable_global = 30

def modificar_variable_global():
    # Uso de una variable global

    variable_global = 40
    print(f"Dentro de la funcin: {variable_global}")

# Llamando a la funcin
modificar_variable_global()

# Imprimiendo la variable global fuera de la funcin
print(f"Fuera de la funcin: {variable_global}")
```

```python
# Scope global modificado mediante palabra clave 'global'

variable_global = 30

def funcion_global():
    global variable_global
    variable_global += 10
    print(f"Variable global dentro de la funcin: {variable_global}")

funcion_global()
print(f"Variable global fuera de la funcin: {variable_global}")
```

Veamos un ejemplo del uso de funciones anidadas y el conocimiento del scope de variables para la modificacin de un diccionario.

```python
# Variables globales para almacenar resultados
resultados = {'Suma': None, 'Resta': None, 'Multiplicacin': None, 'Divisin': None}

def operacion_matematica(a, b):
    # Variable global
    global resultados

    def suma():
        resultados['Suma'] = a + b

    def resta():
        resultados['Resta'] = a - b

    def multiplicacion():
        resultados['Multiplicacin'] = a * b

    def division():
        if b != 0:
            resultados['Divisin'] = a / b
        else:
            resultados['Divisin'] = "No es posible dividir por cero"

    # Llamadas a las funciones anidadas
    suma()
    resta()
    multiplicacion()
    division()

# Llamada a la funcin principal
operacion_matematica(10, 5)

# Imprimir resultados globales
for operacion, resultado in resultados.items():
    print(f"{operacion}: {resultado}")
```

### 1.3. Funciones annimas

---
Funciones annimas

---

Las funciones annimas en Python, tambin conocidas como funciones lambda, son funciones que se definen sin utilizar la palabra clave `def` de manera convencional. Estas funciones se caracterizan por ser simples y expresivas, diseadas para realizar operaciones especficas de manera concisa en un solo lnea de cdigo y son especificadas mediante la palabra clave `lambda`.

Sus caractersticas principales son:

* **Sintaxis Compacta:** La sintaxis de una funcin annima es ms breve que la de una funcin definida con `def`.
La estructura bsica es: `lambda argumentos: expresion`.

* **Sin Nombre Explcito:** A diferencia de las funciones tradicionales, las funciones annimas no tienen un nombre explcito. Se crean y utilizan inmediatamente en el lugar donde se necesitan.

* **Una nica Expresin:** La funcin lambda consiste en una nica expresin, y esta expresin se evala y devuelve automticamente.

```python
# Sintaxis de una funcin lambda

nombre_de_la_funcion = lambda argumentos: expresion

# Es equivalente a
def nombre_de_la_funcion(argumentos):
  expresion
```

```python
# Ejemplo de funcin lambda

suma = lambda x, y: x + y

def suma_2(x,y):
  resultado = x + y
  return resultado

resultado_1 = suma(3, 5)
resultado_2 = suma_2(3, 5)

print(resultado_1)
print(resultado_2)
```

```python
# Funcin lambda condicionada

cuadrado_par = lambda x: x**2 if x % 2 == 0 else None
resultado_par = cuadrado_par(4)

print(resultado_par)
```

```python
# Ejemplo de palabra palndromo

es_palindromo = lambda s: s.lower() == s.lower()[::-1]
resultado1 = es_palindromo("reconocer")
resultado2 = es_palindromo("python")

print(f"'reconocer' es palndromo: {resultado1}")
print(f"'python' es palndromo: {resultado2}")
```

#### 1.4.1. Funcin map()

Funcin map()

---

La funcin `map()` es un mtodo de Python el cul se emplea para aplicar procedimientos como funciones (tanto incluidas dentro de Python, como annimas o construidas por el programador) a una estructura de datos iterable como las listas.

```python
# Sintaxis de la funcin map()

resultado = list( map(funcion_lambda, iterable) )
```

El mtodo `filter()`, como su nombre lo indica, se emplea para filtrar o descartar ciertos elementos de un objeto iterable deacuerdo a un criterio condicional.

# **2. Manejo de errores en Python**

El manejo de errores en Python se refiere a la implementacin de estructuras y tcnicas que permiten gestionar y controlar las excepciones o errores que pueden surgir durante la ejecucin de un programa. Las excepciones son eventos inesperados o condiciones anmalas que pueden ocurrir durante la ejecucin del cdigo y que pueden interrumpir el flujo normal del programa si no se gestionan adecuadamente. Para lidiar con posibles excepciones en el programa se hace uso de las palabras reservadas `try-except`.

La sintaxis bsica de un `try-except` es

```python
# Sintaxis bsica del manejo de errores o excepciones

try:
    # Cdigo propenso a generar excepciones
except ExcepcionTipo1 as e1:
    # Manejar excepcin de tipo 1
except ExcepcionTipo2 as e2:
    # Manejar excepcin de tipo 2
```

Opcionalmente, se pueden utilizar bloques `else` y `finally` despus de los bloques except. El bloque `else` se ejecuta si no se produce ninguna excepcin en el bloque `try`. El bloque `finally` se ejecuta siempre, independientemente de si se produjo una excepcin o no.

```python
# Sintaxis bsica del manejo de errores o excepciones con bloques else y finally

try:
    # Cdigo propenso a generar excepciones
except ExcepcionTipo1 as e1:
    # Manejar excepcin de tipo 1
except ExcepcionTipo2 as e2:
    # Manejar excepcin de tipo 2
else:
    # Cdigo a ejecutar si no hay excepciones
finally:
    # Cdigo que siempre se ejecutar
```

```python
# Ejemplo simple de try-except

a = 13

try:
    b = int(input("Ingrese un numero para dividir a 'a' =  "))
    a = a/b
    print("Su resultado es =",a)
except:
    print("Hay un error!")
```

Un error o excepcion puede ser "bloqueado" mediante una o varias lneas de cdigo. Tamben, los bloques `except` pueden ser programadas para tener en cuenta algn tipo de excepcin especfica o no.

```python
# Ejemplo de mltiples excepciones dentro de un msmo bloque except

try:
    # Cdigo propenso a generar excepciones
except (ZeroDivisionError, NameError):
    # Cdigo a ejecutar si hay excepciones ZeroDivisionError ni NameError

# codigo a ejecutar si no exsten excepciones
```

```python
# Ejemplo de mltiples excepciones programadas explcitamente

try:
    # Cdigo propenso a generar excepciones
except ZeroDivisionError:
    # Cdigo a ejecutar si exste un ZeroDivisionError
except NameError:
    # Cdigo a ejecutar si exste un NameError
except:
    # Cdigo a ejecutar si exste alguna excepcion en general

# codigo a ejecutar si no exsten excepciones
```

Por ejemplo, podrmos ejecutar diferentes bloques de cdigo dependiendo de casos especficos.

```python
# Ejemplo de manejo de excepciones/errores especficas

a = 1

try:
    b = int(input("Ingrese un nmero para dividir a 'a' = "))
    a = a/b
    print("Resultado = ",a)
except ZeroDivisionError:
    print("No es posible realizar la divisin ya que usted ingres a = 0")
except ValueError:
    print("Usted no ingres nmero alguno.")
except:
    print("Algo sali mal, que puede ser?")
```

```python
# Ejemplo de manejo de excepciones/errores especficas con bloque finally

a = 1

try:
    b = int(input("Ingrese un nmero para dividir a 'a' = "))
    a = a/b

except ZeroDivisionError:
    print("No es posible realizar la divisin ya que usted ingres a = 0")
except ValueError:
    print("Usted no ingres nmero alguno.")
except:
    print("Algo sali mal, que puede ser?")
else:
    print("Resultado =",a)
finally:
    print("Felicidades, pudo hacer la divisin")
```

```python
# Ejempo de manejo de excepciones/errores dentro de una funcin

def operacion_compleja(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("Error: Divisin por cero no permitida")
    else:
        print(f"Resultado de la operacin: {resultado}")
    finally:
        print("Este bloque siempre se ejecutar")

# Ejemplo de uso
operacion_compleja(8, 4)
operacion_compleja(6, 0)
```

# **3. Clases y objetos**

En el contexto de la programacin orientada a objetos (OOP), una clase es una plantilla o un "molde" que define la estructura y el comportamiento de objetos. Los objetos son instancias de clases, y las clases proporcionan un mecanismo para organizar y encapsular el cdigo de manera que sea reutilizable y fcil de entender.

**Principales Caractersticas de las Clases:**

* **Atributos:** Las clases pueden tener atributos que representan caractersticas o propiedades del objeto. Estos atributos pueden ser variables que almacenan datos.

* **Mtodos**: Las clases tambin pueden contener mtodos, que son funciones asociadas a la clase y que definen el comportamiento del objeto. Los mtodos operan sobre los atributos del objeto.

* **Encapsulacin**: Las clases permiten encapsular la implementacin interna, ocultando detalles complejos y exponiendo solo la interfaz necesaria para interactuar con el objeto. Esto facilita la modularidad y el mantenimiento del cdigo.

* **Herencia**: La herencia es un concepto clave en OOP que permite que una clase herede atributos y mtodos de otra clase. Esto fomenta la reutilizacin del cdigo y la creacin de jerarquas de clases.

* **Polimorfismo**: El polimorfismo permite que objetos de diferentes clases respondan de manera consistente a un mismo conjunto de mtodos. Esto facilita el uso de interfaces comunes en situaciones diversas.

### 3.1. Definicin de clases, atributos y mtodos

---
Definicin de clases, atributos y mtodos

---

Las clases dentro de Python, se definen como objetos los cuales estn caracterizados por poseer ciertos atributos y mtodos prpios. Una clase puede entenderse como una coleccin de funciones, constantes y tros objetos que se pueden emplear de manera dinmica y generan instancias usables en diferentes contextos.

```python
# Sintaxis bsica

class tipo_de_objeto_o_entidad:
  # Atributos de la clase
  atributo_1 = valor _1
  atributo_2 = valor _2

  def metodo_1(self, parametros):
    expresion

  def metodo_2(self, parametros):
    expresion
```

```python
# Ejemplo bsico de una clase y creacin de un objeto y polimorfsmo

class ave:

    # Atributos de la clase
    nombre = ""
    edad = 0

# Creacin del objeto
ave_1 = ave()
ave_1.nombre = "Blu"
ave_1.edad = 10

# Creacin de otro objeto
ave_2 = ave()
ave_2.nombre = "Cacao"
ave_2.edad = 20

# access attributes
print(f"{ave_1.nombre} tiene {ave_1.edad} aos")
print(f"{ave_2.nombre} tiene {ave_2.edad} aos")
```

```python
# Ejemplo de una clase con mtodos

class Calculadora:
    def sumar(self, a, b):
        return a + b

    def restar(self, a, b):
        return a - b

    def multiplicar(self, a, b):
        return a * b

    def dividir(self, a, b):
        if b != 0:
            return a / b
        else:
            print("Error: Divisin por cero.")

# Crear una instancia de la clase
calculadora = Calculadora()

# Llamar a los mtodos
resultado_suma = calculadora.sumar(5, 3)
resultado_resta = calculadora.restar(5, 3)
resultado_multiplicacion = calculadora.multiplicar(5, 3)
resultado_division = calculadora.dividir(5, 3)

# Imprimir resultados
print(f"Suma: {resultado_suma}")
print(f"Resta: {resultado_resta}")
print(f"Multiplicacin: {resultado_multiplicacion}")
print(f"Divisin: {resultado_division}")
```

### 3.1. Encapsulamiento

---
Encapsulamiento

La encapsulacin es una de las caractersticas clave de la programacin orientada a objetos. La encapsulacin consiste en agrupar atributos y mtodos dentro de una misma clase. Evita que las clases externas accedan a los atributos y mtodos de una clase y los modifiquen. Esto tambin ayuda a lograr la ocultacin de datos. En Python, denotamos atributos privados utilizando el guin bajo como prefijo, es decir,` _` o `__`.

* `__init__`:
  * Es un mtodo especial en Python que se llama automticamente cuando se crea una nueva instancia de una clase.
  * Este mtodo se utiliza para inicializar los atributos de la clase o realizar otras acciones de configuracin necesarias al crear un objeto.
  * Se usa para asignar valores iniciales a los atributos de la clase y puede aceptar argumentos que se pasan al crear una instancia.

* `__variable`:

  * La convencin de nombres con doble guion bajo (`__`) se utiliza para indicar que una variable es privada y no debe ser accedida directamente desde fuera de la clase.
  * Aunque en Python no hay restricciones estrictas de acceso, la convencin `__variable` sugiere que la variable es para uso interno y no debe ser modificada externamente.
  * La encapsulacin en Python se basa en esta convencin y la buena prctica de no acceder directamente a variables privadas desde fuera de la clase.

La definicin de una clase con encapsulacin implica el uso de palabras clave tales como `class` y `__init__()`, veamos unos ejemplos.

```python
# Sintaxis bsica de una clase

class un_tipo_de_objeto:
  def __init__(parametros o atributos):        # Esta lnea de cdigo inicializa la clase
    self.atributo_intrnseco_al_objeto_1 = valor_1
    self.atributo_intrnseco_al_objeto_2 = valor_2

  def metodo_del_objeto(parametros):
    expresion
```

En Python, la palabra reservada `self` se utiliza como convencin para referirse al objeto actual de una clase. En el contexto de las clases y objetos, `self `se utiliza como el primer parmetro de los mtodos de una clase y representa la instancia de la propia clase.

```python
# Ejemplo de definicin de una clase con encapsulamiento

class Persona:
    def __init__(self, nombre, edad):
        self.__nombre = nombre  # Atributo privado
        self.__edad = edad      # Atributo privado

    def obtener_nombre(self):
        return self.__nombre

    def obtener_edad(self):
        return self.__edad

    def imprimir_informacion(self):
        print(f"Nombre: {self.__nombre}, Edad: {self.__edad}")

# Crear una instancia de la clase
persona1 = Persona("Alice", 25)

# Intentar acceder directamente a los atributos privados (no es recomendado)
# print(persona1.__nombre)  # Esto generara un error

# Acceder a travs de mtodos
print(f"Nombre: {persona1.obtener_nombre()}, Edad: {persona1.obtener_edad()}")

# Imprimir informacin utilizando un mtodo pblico
persona1.imprimir_informacion()
```

```python
# Ejemplo de definicin de una clase con encapsulamiento

class Computador:

    def __init__(self):
        self.__preciomaximo = 900

    def venta(self):
        print("Precio de venta: {}".format(self.__preciomaximo))

    def FijarPrecioMaximo(self, precio):
        self.__preciomaximo = precio

c = Computador()
c.venta()

# Intento de cambio de precio
c.__preciomaximo = 1000
c.venta()

# Usar una funcin que tome como argumento el atributo privado para cambiar el atributo privado
c.FijarPrecioMaximo(1000)
c.venta()
```

```python
# Ejemplo de una clase como un objeto usado para procesar datos

class ProcesadorNumeros:
    def __init__(self, numeros):
        self.numeros = numeros

    def sumar_elementos(self):
        suma = 0
        for numero in self.numeros:
            suma += numero
        return suma

# Crear una instancia de la clase
numeros_a_procesar = [1, 2, 3, 4, 5]
procesador = ProcesadorNumeros(numeros_a_procesar)

# Llamar al mtodo para sumar elementos
resultado_suma = procesador.sumar_elementos()

# Imprimir el resultado
print(f"La suma de los elementos es: {resultado_suma}")
```

```python
# Ejemplo de una clase Calculadora con encapsulamiento

class Calculadora:
    def __init__(self, numero1, numero2):
        self.__numero1 = numero1
        self.__numero2 = numero2

    def sumar(self):
        return self.__numero1 + self.__numero2

    def restar(self):
        return self.__numero1 - self.__numero2

    def multiplicar(self):
        return self.__numero1 * self.__numero2

    def dividir(self):
        if self.__numero2 != 0:
            return self.__numero1 / self.__numero2
        else:
            print("Error: Divisin por cero.")
            return None

# Crear una instancia de la clase
calculadora = Calculadora(10, 5)

# Llamar a los mtodos encapsulados
resultado_suma = calculadora.sumar()
resultado_resta = calculadora.restar()
resultado_multiplicacion = calculadora.multiplicar()
resultado_division = calculadora.dividir()

# Imprimir resultados
print(f"Suma: {resultado_suma}")
print(f"Resta: {resultado_resta}")
print(f"Multiplicacin: {resultado_multiplicacion}")
print(f"Divisin: {resultado_division}")
```

La herencia es un mecanismo en programacin orientada a objetos donde una clase puede adquirir atributos y mtodos de otra clase. La clase derivada hereda las caractersticas de la clase base, lo que facilita la reutilizacin del cdigo y la creacin de una jerarqua de clases.

```python
# Sintaxis bsica de herencia de clases

class clase_1:
  atributo_1 = valor
  atributo_2 = valor

   def metodo_1(self):
    expresion

  def metodo_2(self):
    expresion

class clase_2(clase_2): # Clase que hereda los metodos de otra clase
  atributo_1 = valore
  atributo_2 = valor

  def metodo_1(self):
    expresion
```

```python
# Ejemplo de herencia simple

class persona:

  def nombre(self):
    print('Mario')

  def apellido(self):
    print('Rodriguez')

class otra_persona(persona):
  def comentario(self):
    print('Puedo hacer comentarios!')

p = otra_persona()

p.nombre()
p.apellido()
p.comentario()
```

```python
# Ejemplo de clase cvehculo

class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def mostrar_informacion(self):
        print(f"Marca: {self.marca}, Modelo: {self.modelo}")

class Coche(Vehiculo):
    def __init__(self, marca, modelo, color):
        # Llamada al constructor de la clase base
        super().__init__(marca, modelo)
        self.color = color

    def mostrar_informacion_coche(self):
        print(f"Color: {self.color}")

# Crear una instancia de la clase derivada (Coche)
mi_coche = Coche("Toyota", "Camry", "Azul")

# Acceder a los mtodos de la clase base y la clase derivada
mi_coche.mostrar_informacion()  # Mtodos de la clase base
mi_coche.mostrar_informacion_coche()  # Mtodo de la clase derivada
```
