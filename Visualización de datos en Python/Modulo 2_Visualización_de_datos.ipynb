# **1. Barplots e Histogramas**

---
Introduccin a Seaborn

---

Seaborn, creada por Michael Waskom, se presenta como una biblioteca de visualizacin de datos en Python que ampla y mejora las capacidades de Matplotlib. Su desarrollo comenz en 2012 como una respuesta a la necesidad de simplificar y embellecer las visualizaciones estadsticas en Python. Desde entonces, Seaborn ha evolucionado, convirtindose en una herramienta esencial en el arsenal de analistas de datos y cientficos.

* **Esttica Visual Sofisticada:**
Seaborn destaca por su enfoque en la esttica visual. Ofrece paletas de colores atractivas y esquemas de colores ms refinados, lo que contribuye a la creacin de grficos visualmente impactantes y profesionales.

* **Sintaxis Simplificada:**
A diferencia de Matplotlib, Seaborn presenta una sintaxis simplificada que facilita la creacin de visualizaciones complejas con menos lneas de cdigo. Se integra naturalmente con estructuras de datos como DataFrames de Pandas, simplificando la manipulacin y visualizacin de datos.

* **Grficos Estadsticos Integrados:**
Seaborn ofrece una variedad de grficos estadsticos integrados que simplifican la exploracin de relaciones entre variables. Funciones como `lmplot`, `pairplot` y `jointplot` permiten visualizar patrones y correlaciones de manera intuitiva.

* **Facilidad de Personalizacin:**
Aunque Seaborn se basa en Matplotlib, aade una capa adicional de abstraccin que facilita la personalizacin rpida y eficiente de grficos. Las opciones de estilo y los parmetros visuales son accesibles con facilidad, permitiendo adaptar la apariencia de los grficos segn las necesidades especficas.

* **Soporte para Datos Categricos:**
Seaborn sobresale en la visualizacin de datos categricos. Funciones como catplot y boxplot proporcionan herramientas poderosas para explorar relaciones entre variables categricas.

* **Integracin con Pandas y NumPy:**
Seaborn se integra sin esfuerzo con otras bibliotecas fundamentales de Python, como Pandas y NumPy. Esta interoperabilidad facilita la transicin fluida entre manipulacin de datos y creacin de visualizaciones.

En Seaborn, `sns.set(style,palette,font_scale)` se utiliza para establecer ciertos aspectos visuales por defecto en todas las visualizaciones subsiguientes que realices con Seaborn. Cada uno de los argumentos en este comando tiene un propsito especfico.

### 1.1. Barplots

---
Barplots

---

Dentro de los plots ms sencillos que se pueden encontar en Seaborn, est el comando `.barplot()` cuya sintaxis es `sns.barplot(x, y, hue, data, estimator, ci, n_boot, units, seed, orient, color, palette, saturation, errcolor, errwidth, capsize, dodge, ax, **kwargs)` en donde:

* `x`: Variable categrica que se colocar en el eje x.
* `y`: Variable numrica cuyos valores se utilizarn para la altura de las barras.
* `hue`: Variable categrica adicional que se utilizar para segmentar los datos y agregar una dimensin adicional a la visualizacin.
* `data`: DataFrame que contiene los datos.
* `estimator`: Funcin utilizada para estimar la altura de las barras. Por defecto, es la media (<function mean>), pero puede cambiarse segn las necesidades.
* `ci`: Nivel de confianza para los intervalos de confianza alrededor de la estimacin.
palette: Paleta de colores a utilizar en el grfico.
saturation: Controla la saturacin del color.
* `errcolor`: Color de las lneas de error.
* `errwidth`: Ancho de las lneas de error.
* `capsize`: Ancho de las "tapas" en las lneas de error.
* `dodge`: Si es True, agrega barras separadas para cada nivel de la variable hue.
* `ax`: Eje en el cual dibujar el grfico.

```python
# Barplot simple
sns.barplot(x=['A','B','C'],
            y=[1,4.5,2.7],
            palette='muted')
plt.show()
```

```python
# Crear un grfico de barras con Seaborn
sns.barplot(x='day', y='total_bill',
            data=tips_data,
            estimator=np.mean, ci=95,
            palette='muted')

# Personalizar el grfico
plt.title('Promedio de Factura Total por Da')
plt.xlabel('Da de la Semana')
plt.ylabel('Factura Total Promedio')
plt.show()
```

### 1.2. Histogramas

---
Histogramas

---

La funcin `sns.displot` en Seaborn es una herramienta verstil diseada para explorar y visualizar distribuciones univariadas de datos. A travs de histogramas, grficos de densidad y otros componentes, `displot` facilita la comprensin de la forma y la estructura de los datos, revelando patrones fundamentales que pueden guiar el anlisis exploratorio de datos. Su sintaxis bsica es `sns.displot(data, x, y, hue, kind, rug, col, row, palette, aspect, height, facet_kws, **kwargs)`

* `data`: El DataFrame que contiene los datos.
* `x, y`: Variables a representar en los ejes x e y. Al menos una de ellas debe ser especificada.
* `hue`: Variable categrica que se utilizar para colorear los elementos en el grfico.
* `kind`: Tipo de grfico a crear ('hist', 'kde', 'ecdf', 'rug'). Por defecto, es 'hist'.
* `rug`: Un booleano que indica si se deben agregar marcas en el eje x.
* `col, row`: Variables categricas que dividen los datos en subconjuntos y crean grficos por facetas.
* `palette`: Paleta de colores a utilizar en el grfico.
* `aspect`: Relacin de aspecto de cada grfico en las facetas.
* `height`: Altura de cada grfico en las facetas.
* `facet_kws`: Argumentos adicionales para personalizar las facetas.



```python
# Hostograma de una variable
plt.figure(figsize=(4,3))
sns.displot(data=tips_data, x='total_bill')
plt.show()
```

```python
# Cargar datos de ejemplo
tips_data = sns.load_dataset('tips')

# Crear un displot para la variable 'total_bill'
sns.displot(tips_data, x='total_bill',
            kde=True, bins=30, color='blue', rug=True)

# Personalizar el grfico
plt.title('Distribucin de Factura Total')
plt.xlabel('Factura Total')
plt.ylabel('Frecuencia')
plt.show()
```

En Seaborn, las facetas en sns.displot se refieren a la capacidad de dividir un grfico en mltiples subgrficos (facetas) segn los niveles de una o ms variables categricas. Esta funcin facilita la exploracin de las distribuciones condicionales o comparativas de las variables en los subconjuntos de datos definidos por las categoras.

```python
# Crear un displot con facetas por la variable 'day'
sns.displot(data=tips_data, x='total_bill',
            kde=True, bins=30,
            color='blue', rug=True,
            col='day')
plt.show()
```

El `hue` se utiliza para comparar las distribuciones de una variable entre los niveles de otra variable categrica. Cada grupo se representa en un color diferente, lo que facilita la identificacin de diferencias en las distribuciones entre las variables.

```python
# Uso del parmetro hue
plt.figure(figsize=(4,2))
sns.displot(data=tips_data, x='total_bill',
            hue='day',
            kind='kde',
            legend=True,#palette='dark',
            alpha=0.9)
plt.show()
```

```python
# Uso del parmetro hue
plt.figure(figsize=(4,2))
sns.histplot(data=tips_data,
             x='total_bill',
             hue='sex')
plt.show()
```

# **2. Distribuciones**

---
Distribuciones y subplots

---

```python
sns.set(style='ticks',palette='dark')
tips = sns.load_dataset('tips')
tips .head()
```

La funcin `sns.histplot` en Seaborn es una herramienta especializada para crear histogramas, que son grficos que muestran la frecuencia de ocurrencia de valores en intervalos especficos. Este mtodo simplifica la exploracin de la distribucin de una variable numrica al representar de manera visual la concentracin de datos en diferentes rangos.

```python
# Crear un histograma con Seaborn
sns.histplot(tips, x='total_bill', bins=30,
             kde=True, color='blue')
plt.show()
```

```python
# Crear un histograma con Seaborn y uso del parmetro hue
plt.figure(figsize=(4,3))
sns.displot(data=tips,x='tip',hue='sex',
            stat='count',multiple='stack',kde=True)
plt.show()
```

Para crear un subplot usando el comando `histplot` en conjunto con la visualizacines basadas en programacin orientada a objetos, dentro de Seaborn podemos especificar el eje donde el grfico ha de estar especificando dcha posicin en el parmetro `ax` de cada visual.

```python
# Uso de subplots e histogramas
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6, 4))
sns.histplot(data=tips, x='total_bill', y='tip',
             ax=axes[0])
sns.histplot(data=tips, x='total_bill', y='tip', hue='sex',
             ax=axes[1])
plt.subplots_adjust(wspace=0.4)
axes[0].set_title('Histograma 1')
axes[1].set_title('Histograma 2')
plt.show()
```

Dentro de `histplot`, el parmetro `stat` controla la estadstica que se visualiza en el eje del hisograma. Algunos de ellos son:

* `'count'`: Muestra la cantidad de observaciones en cada bin del histograma.
* `'frequency'`: Muestra la frecuencia relativa de observaciones en cada bin. La altura de cada barra representa la proporcin del total de observaciones.
* `'density'`: Normaliza la frecuencia relativa para qeu la intrgral del histograma sea igual a 1. La altura de cada barra representa la densidad de probabilidad.
* `'probability'`: Similar a `'density'`, normaliza la frecuencia relativa para que la integral del histograma sea igual a 1.
* `'percent o percentile'`: Muestra el porcentaje acumularo de obsevaciones en cada bin
* `'cumulative`: Muestra el nmero acumulado de observaciones en cada bin.

```python
# Matriz de hostogramas en seaborn
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
sns.histplot(data=tips,x='tip',
             bins=15,hue='sex',
             stat='probability',
             ax=axes[0,0])

sns.histplot(data=tips,x='tip',
             bins=15,hue='sex',
             cumulative=True,
             stat='probability',
             ax=axes[0,1])

sns.histplot(data=tips,x='total_bill',
             bins=15,hue='sex',
             stat='percent',
             ax=axes[1,0])

sns.histplot(data=tips,x='total_bill',
             bins=15,hue='sex',
             cumulative=True,
             stat='percent',
             ax=axes[1,1])

plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

La funcin `sns.kdeplot` en Seaborn se utiliza para visualizar estimaciones de densidad del kernel (KDE), que proporcionan una representacin suavizada de la distribucin de una variable continua. Este mtodo es especialmente til para comprender la forma y la estructura de la distribucin, ya que suaviza la irregularidad inherente a los histogramas. Su sintaxis bsica es `sns.kdeplot(data['variable'], shad, color)` en donde:

* `data['variable']`: La variable continua que se desea visualizar.
* `shade`: Un booleano que controla si el rea bajo la curva KDE se sombrea.
* `color`: Define el color de la curva KDE.

```python
# Matriz de grficos kde
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
sns.kdeplot(data=tips,x='tip',
            hue='sex',
            ax=axes[0,0])
sns.kdeplot(data=tips,x='tip',
            hue='sex',cumulative=True,
            ax=axes[0,1])
sns.kdeplot(data=tips,x='tip',
            hue='sex',shade=True,
            ax=axes[1,0])
sns.kdeplot(data=tips,x='tip',
            hue='sex',cumulative=True,
            shade=True,
            ax=axes[1,1])
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

La funcin `sns.ecdfplot` en Seaborn se utiliza para visualizar funciones de distribucin acumulativa emprica (ECDF), que muestran cmo se acumulan los valores en una distribucin. La ECDF es una herramienta til para entender la distribucin de datos y la probabilidad asociada a diferentes valores. Su sintaxis bsica es `sns.ecdfplot(data['variable'], stat, complementary, color)`. En donde:

* `data['variable']`: La variable continua que se desea visualizar.
* `stat`: Controla el tipo de ECDF que se visualiza. Puede ser 'proportion', 'percentile' o 'kde'.
* `complementary`: Un booleano que indica si mostrar la ECDF complementaria.
* `color`: Define el color de la ECDF.

```python
# Funcin acumulativa
plt.figure(figsize=(6,3))
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

#**3. Plots para variables categricas**

---
Plots para variables categricas

---

La visualizacin de datos categricos es esencial para comprender la distribucin y las relaciones entre diferentes grupos en un conjunto de datos. Seaborn, una biblioteca de visualizacin basada en Matplotlib, proporciona herramientas poderosas y atractivas para explorar variables categricas de manera efectiva.

```python
import seaborn as sns
sns.load_dataset('tips')
tips.head()
```

La funcin `sns.countplot` en Seaborn se utiliza para visualizar la distribucin de frecuencias de una variable categrica. Es especialmente til para contar la cantidad de observaciones en cada categora y obtener una visin rpida de la proporcin de cada clase en un conjunto de datos. Su sintaxis bsica es `sns.countplot(x='variable', data, hue, palette, order, orient)`, en donde:

* `x`: La variable categrica que se desea visualizar en el eje x.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categrica, creando barras agrupadas o apiladas segn los niveles de esta variable.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `order`: Opcional. Especifica el orden en el que se deben mostrar las categoras en el eje x.
* `orient`: Opcional. Puede ser 'v' (vertical) o 'h' (horizontal), controlando la orientacin de las barras.



```python
# Countplot bsico
plt.figure(figsize=(4,3))
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

La funcin `sns.stripplot` en Seaborn se utiliza para visualizar puntos individuales en una lnea para cada categora de una variable categrica. Es particularmente til cuando se desea mostrar la distribucin de datos y la densidad de puntos en un conjunto de categoras. Su sintaxis bsica es `sns.stripplot(x='variable', y='value', data, hue, palette, dodge, jitter, marker)` en donde:

* `x`: La variable categrica que se desea visualizar en el eje x.
* `y`: La variable numrica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categrica, creando tiras separadas para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `dodge`: Opcional. Si es True, separa las tiras para cada nivel de la variable hue.
* `jitter`: Opcional. Controla la dispersin horizontal de los puntos para evitar superposiciones.
* `marker`: Opcional. Especifica el estilo del marcador para los puntos en el strip plot.


```python
# Strip plot bsico con separacin de hue
plt.figure(figsize=(4,3))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en funcin del sa y del sexo')
plt.show()
```

La funcin `sns.swarmplot` en Seaborn se utiliza para visualizar puntos individuales en una lnea para cada categora de una variable categrica. Al igual que `sns.stripplot`, es til para mostrar la distribucin de datos y la densidad de puntos en un conjunto de categoras. Sin embargo, a diferencia de `stripplot`, `swarmplot` evita que los puntos se superpongan, proporcionando una visin ms clara de la distribucin. Su sintaxis bsica es `sns.swarmplot(x='variable', y='value', data, hue, palette, dodge, size, edgecolor)` en donde:

* `x`: La variable categrica que se desea visualizar en el eje x.
* `y`: La variable numrica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categrica, creando enjambres separados para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `dodge`: Opcional. Si es True, separa los enjambres para cada nivel de la variable hue.
* `size`: Opcional. Controla el tamao de los puntos en el enjambre.
* `edgecolor`: Opcional. Define el color del borde de los puntos en el enjambre.


```python
# Swarmplot plot bsico con separacin de hue
plt.figure(figsize=(4,3))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en funcin del sa y del sexo')
plt.show()
```

La funcin `sns.boxplot` en Seaborn se utiliza para visualizar la distribucin de una variable numrica en diferentes niveles de una variable categrica. Proporciona informacin sobre la mediana, cuartiles y posibles valores atpicos en cada categora, lo que facilita la comparacin de distribuciones. Su sintaxis bsica es `sns.boxplot(x='variable_categorica', y='variable_numerica', data, hue, palette, order)`, en donde:

* `x`: La variable categrica que se desea visualizar en el eje x.
* `y`: La variable numrica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categrica, creando cajas separadas para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `order`: Opcional. Especifica el orden en el que se deben mostrar las categoras en el eje x.

```python
# Boxplot plot bsico con separacin de hue
plt.figure(figsize=(4,3))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en funcin del sa y del sexo')
plt.show()
```

```python
# Combinacin de swarmplot y boxplot
plt.figure(figsize=(4,3))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,
              color='0',marker='<')
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en funcin del sa y del sexo')
plt.show()
```

La funcin `sns.violinplot` en Seaborn se utiliza para visualizar la distribucin de una variable numrica en diferentes niveles de una variable categrica. A diferencia de un box plot, violinplot combina la informacin de un box plot con una estimacin de densidad del kernel (KDE), proporcionando una representacin ms completa de la distribucin. Su sintaxis bsica es `sns.violinplot(x='variable_categorica', y='variable_numerica', data, hue, palette, split, inner, bw) en donde:

* `x`: La variable categrica que se desea visualizar en el eje x.
* `y`: La variable numrica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categrica, creando violines separados para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `split`: Opcional. Si es True, divide los violines cuando hay una variable categrica en el eje x.
* `inner`: Opcional. Controla la representacin interna del violin, como 'box' para mostrar un box plot en su interior.
* `bw`: Opcional. Controla la anchura del kernel en la estimacin de densidad.

```python
# Combinacin de violinplots, boxplots y swarmplot
figure, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,4))
sns.violinplot(data=tips,x='day',y='total_bill',
               hue='sex',dodge=True,ax=axes[0])
sns.swarmplot(data=tips,x='day',y='total_bill',
              hue='sex',dodge=True,color='0',marker='<',
              ax=axes[0])
sns.violinplot(data=tips,x='day',y='total_bill',
               hue='sex',dodge=False,split=False,
               ax=axes[1])

plt.subplots_adjust(wspace=0.4)
axes[0].set_ylabel('Total bill')
axes[1].set_ylabel('Total bill')
axes[0].set_xlabel('Day')
axes[1].set_xlabel('Day')
axes[0].set_title('Tips as a function of the day and sex')
axes[1].set_title('Tips as a function of the day and sex')
plt.show()
```

La funcin `sns.catplot` en Seaborn es una herramienta verstil para visualizar variables categricas y numricas en diferentes tipos de plots categricos. catplot permite explorar patrones y relaciones en datos complejos mediante la creacin de subplots basados en la estructura de un conjunto de datos. Su sintaxis bsica es `sns.catplot(x='variable_categorica', y='variable_numerica', data, kind, hue, palette, col, row), en donde:

* `x`: La variable categrica que se desea visualizar en el eje x.
* `y`: La variable numrica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `kind`: Tipo de plot categrico a utilizar ('strip', 'swarm', 'box', 'violin', etc.).
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categrica.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `col y row`: Opcionales. Permiten segmentar el plot en columnas y filas segn variables categricas adicionales.

```python
# Catplot bsico
sns.catplot(data=tips,x='day',y='total_bill',
            hue='sex',dodge=True,kind='box',
            col='time')
plt.show()
```

# **4. Plots para relaciones entre variables**

Plots para relaciones entre variables

La visualizacin de relaciones entre variables es esencial en el anlisis de datos, ya que permite comprender patrones, tendencias y posibles correlaciones. Seaborn ofrece herramientas poderosas para crear grficos que destacan las relaciones entre diferentes variables. A continuacin, se presenta una introduccin a algunos de los plots ms utilizados para explorar relaciones en datos.

```python
tips_data=sns.load_dataset('tips')
```

La funcin `sns.scatterplot` en Seaborn se utiliza para crear scatter plots, que son grficos que muestran la relacin entre dos variables numricas mediante la colocacin de puntos en un plano cartesiano. Este tipo de visualizacin es valioso para identificar patrones, tendencias y posibles correlaciones entre variables. Acepta parmetros como:

* `x`: Especifica la variable que se mapear que en el eje x.
* `y`: Especifica la variable que se mapear en el eje y.
* `data`: DataFrame que contiene los datos.
* `hue`: Permite segmentar los datos por una tercera variable categrica, coloreando los puntos.
* `style`: Permite segmentar los datos por una cuarta variable categrica, usando diferentes estilos de marcador.
* `size`: Permite segmentar los datos por una quinta variable numrica, ajustando el tamao de los marcadores.

```python
# Scatter plot con mltiples divisiones
sns.scatterplot(data=tips_data,x='total_bill',y='tip',
                hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.15,0.5))
plt.grid(True)
plt.show()
```

```python
# Scatter plot con mltiples divisiones y markers definidos
plt.figure(figsize=(8,8))
markers = {'Lunch':'D','Dinner':'s'}
sns.scatterplot(data=tips_data,x='total_bill',y='tip',
                hue='day',style='time',size='size',
                markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.15,0.5))
plt.grid(True)
plt.show()
```

Este mtodo es especialmente til para visualizar la relacin entre variables en conjuntos de datos ms grandes. * `y`: Variable que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.

La funcin `sns.pairplot` en Seaborn se utiliza para crear una matriz de grficos que muestra las relaciones bivariadas entre mltiples variables en un conjunto de datos. # **5. Heatmaps**

La funcin `sns.heatmap` en Seaborn se utiliza para crear mapas de calor, que son visualizaciones efectivas para representar matrices de datos bidimensionales. Es especialmente til para resaltar patrones y tendencias en datos tabulares. Su sintaxis bsica es `sns.heatmap(data, annot, cmap, linewidths, cbar)`, en donde:

* `data` : La matriz bidimensional o el DataFrame que se desea visualizar.
* `annot` : Si es `True`, muestra los valores dentro de cada celda.
* `cmap` : Especifica la paleta de colores a utilizar.
* `linewidths` : Ancho de las lneas divisorias entre celdas.
* `cbar` : Si es `True`, muestra una barra de color que indica la convencin de color empleada.

```python
# Heatmap de la correlacin entre variables
cor = tips_data.corr()
sns.heatmap(cor,annot=True, cmap='coolwarm',linewidths=5,
            linecolor='black', vmin=0.4, vmax=1,
            cbar=True)
plt.show()
```
