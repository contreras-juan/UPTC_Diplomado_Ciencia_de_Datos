# **1. Barplots e Histogramas**

---
Introducción a Seaborn

---

Seaborn, creada por Michael Waskom, se presenta como una biblioteca de visualización de datos en Python que amplía y mejora las capacidades de Matplotlib. Su desarrollo comenzó en 2012 como una respuesta a la necesidad de simplificar y embellecer las visualizaciones estadísticas en Python. Desde entonces, Seaborn ha evolucionado, convirtiéndose en una herramienta esencial en el arsenal de analistas de datos y científicos.

* **Estética Visual Sofisticada:**
Seaborn destaca por su enfoque en la estética visual. Ofrece paletas de colores atractivas y esquemas de colores más refinados, lo que contribuye a la creación de gráficos visualmente impactantes y profesionales.

* **Sintaxis Simplificada:**
A diferencia de Matplotlib, Seaborn presenta una sintaxis simplificada que facilita la creación de visualizaciones complejas con menos líneas de código. Se integra naturalmente con estructuras de datos como DataFrames de Pandas, simplificando la manipulación y visualización de datos.

* **Gráficos Estadísticos Integrados:**
Seaborn ofrece una variedad de gráficos estadísticos integrados que simplifican la exploración de relaciones entre variables. Funciones como `lmplot`, `pairplot` y `jointplot` permiten visualizar patrones y correlaciones de manera intuitiva.

* **Facilidad de Personalización:**
Aunque Seaborn se basa en Matplotlib, añade una capa adicional de abstracción que facilita la personalización rápida y eficiente de gráficos. Las opciones de estilo y los parámetros visuales son accesibles con facilidad, permitiendo adaptar la apariencia de los gráficos según las necesidades específicas.

* **Soporte para Datos Categóricos:**
Seaborn sobresale en la visualización de datos categóricos. Funciones como catplot y boxplot proporcionan herramientas poderosas para explorar relaciones entre variables categóricas.

* **Integración con Pandas y NumPy:**
Seaborn se integra sin esfuerzo con otras bibliotecas fundamentales de Python, como Pandas y NumPy. Esta interoperabilidad facilita la transición fluida entre manipulación de datos y creación de visualizaciones.

En Seaborn, `sns.set(style,palette,font_scale)` se utiliza para establecer ciertos aspectos visuales por defecto en todas las visualizaciones subsiguientes que realices con Seaborn. Cada uno de los argumentos en este comando tiene un propósito específico.

```python
sns.set(style='darkgrid',palette='muted',font_scale=1)
```

### 1.1. Barplots

---
Barplots

---

Dentro de los plots más sencillos que se pueden encontar en Seaborn, está el comando `.barplot()` cuya sintaxis es `sns.barplot(x, y, hue, data, estimator, ci, n_boot, units, seed, orient, color, palette, saturation, errcolor, errwidth, capsize, dodge, ax, **kwargs)` en donde:

* `x`: Variable categórica que se colocará en el eje x.
* `y`: Variable numérica cuyos valores se utilizarán para la altura de las barras.
* `hue`: Variable categórica adicional que se utilizará para segmentar los datos y agregar una dimensión adicional a la visualización.
* `data`: DataFrame que contiene los datos.
* `estimator`: Función utilizada para estimar la altura de las barras. Por defecto, es la media (<function mean>), pero puede cambiarse según las necesidades.
* `ci`: Nivel de confianza para los intervalos de confianza alrededor de la estimación.
palette: Paleta de colores a utilizar en el gráfico.
saturation: Controla la saturación del color.
* `errcolor`: Color de las líneas de error.
* `errwidth`: Ancho de las líneas de error.
* `capsize`: Ancho de las "tapas" en las líneas de error.
* `dodge`: Si es True, agrega barras separadas para cada nivel de la variable hue.
* `ax`: Eje en el cual dibujar el gráfico.

```python
# Barplot simple
sns.barplot(x=['A','B','C'],
            y=[1,4.5,2.7],
            palette='muted')
plt.show()
```

```python
# Crear un gráfico de barras con Seaborn
sns.barplot(x='day', y='total_bill',
            data=tips_data,
            estimator=np.mean, ci=95,
            palette='muted')

# Personalizar el gráfico
plt.title('Promedio de Factura Total por Día')
plt.xlabel('Día de la Semana')
plt.ylabel('Factura Total Promedio')
plt.show()
```

### 1.2. Histogramas

---
Histogramas

---

La función `sns.displot` en Seaborn es una herramienta versátil diseñada para explorar y visualizar distribuciones univariadas de datos. A través de histogramas, gráficos de densidad y otros componentes, `displot` facilita la comprensión de la forma y la estructura de los datos, revelando patrones fundamentales que pueden guiar el análisis exploratorio de datos. Su sintaxis básica es `sns.displot(data, x, y, hue, kind, rug, col, row, palette, aspect, height, facet_kws, **kwargs)`

* `data`: El DataFrame que contiene los datos.
* `x, y`: Variables a representar en los ejes x e y. Al menos una de ellas debe ser especificada.
* `hue`: Variable categórica que se utilizará para colorear los elementos en el gráfico.
* `kind`: Tipo de gráfico a crear ('hist', 'kde', 'ecdf', 'rug'). Por defecto, es 'hist'.
* `rug`: Un booleano que indica si se deben agregar marcas en el eje x.
* `col, row`: Variables categóricas que dividen los datos en subconjuntos y crean gráficos por facetas.
* `palette`: Paleta de colores a utilizar en el gráfico.
* `aspect`: Relación de aspecto de cada gráfico en las facetas.
* `height`: Altura de cada gráfico en las facetas.
* `facet_kws`: Argumentos adicionales para personalizar las facetas.



```python
# Hostograma de una variable
plt.figure(figsize=(4,3))
sns.displot(data=tips_data, x='total_bill')
plt.show()
```

```python
# Cargar datos de ejemplo
tips_data = sns.load_dataset('tips')

# Crear un displot para la variable 'total_bill'
sns.displot(tips_data, x='total_bill',
            kde=True, bins=30, color='blue', rug=True)

# Personalizar el gráfico
plt.title('Distribución de Factura Total')
plt.xlabel('Factura Total')
plt.ylabel('Frecuencia')
plt.show()
```

En Seaborn, las facetas en sns.displot se refieren a la capacidad de dividir un gráfico en múltiples subgráficos (facetas) según los niveles de una o más variables categóricas. Esta función facilita la exploración de las distribuciones condicionales o comparativas de las variables en los subconjuntos de datos definidos por las categorías.

```python
# Crear un displot con facetas por la variable 'day'
sns.displot(data=tips_data, x='total_bill',
            kde=True, bins=30,
            color='blue', rug=True,
            col='day')
plt.show()
```

El `hue` se utiliza para comparar las distribuciones de una variable entre los niveles de otra variable categórica. Cada grupo se representa en un color diferente, lo que facilita la identificación de diferencias en las distribuciones entre las variables.

```python
# Uso del parámetro hue
plt.figure(figsize=(4,2))
sns.displot(data=tips_data, x='total_bill',
            hue='day',
            kind='kde',
            legend=True,#palette='dark',
            alpha=0.9)
plt.show()
```

```python
# Uso del parámetro hue
plt.figure(figsize=(4,2))
sns.histplot(data=tips_data,
             x='total_bill',
             hue='sex')
plt.show()
```

# **2. Distribuciones**

---
Distribuciones y subplots

---

```python
sns.set(style='ticks',palette='dark')
tips = sns.load_dataset('tips')
tips .head()
```

La función `sns.histplot` en Seaborn es una herramienta especializada para crear histogramas, que son gráficos que muestran la frecuencia de ocurrencia de valores en intervalos específicos. Este método simplifica la exploración de la distribución de una variable numérica al representar de manera visual la concentración de datos en diferentes rangos.

```python
# Crear un histograma con Seaborn
sns.histplot(tips, x='total_bill', bins=30,
             kde=True, color='blue')
plt.show()
```

```python
# Crear un histograma con Seaborn y uso del parámetro hue
plt.figure(figsize=(4,3))
sns.displot(data=tips,x='tip',hue='sex',
            stat='count',multiple='stack',kde=True)
plt.show()
```

Para crear un subplot usando el comando `histplot` en conjunto con la visualizaciónes basadas en programación orientada a objetos, dentro de Seaborn podemos especificar el eje donde el gráfico ha de estar especificando dícha posición en el parámetro `ax` de cada visual.

```python
# Uso de subplots e histogramas
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6, 4))
sns.histplot(data=tips, x='total_bill', y='tip',
             ax=axes[0])
sns.histplot(data=tips, x='total_bill', y='tip', hue='sex',
             ax=axes[1])
plt.subplots_adjust(wspace=0.4)
axes[0].set_title('Histograma 1')
axes[1].set_title('Histograma 2')
plt.show()
```

Dentro de `histplot`, el parámetro `stat` controla la estadística que se visualiza en el eje del hisograma. Algunos de ellos son:

* `'count'`: Muestra la cantidad de observaciones en cada bin del histograma.
* `'frequency'`: Muestra la frecuencia relativa de observaciones en cada bin. La altura de cada barra representa la proporción del total de observaciones.
* `'density'`: Normaliza la frecuencia relativa para qeu la intrgral del histograma sea igual a 1. La altura de cada barra representa la densidad de probabilidad.
* `'probability'`: Similar a `'density'`, normaliza la frecuencia relativa para que la integral del histograma sea igual a 1.
* `'percent o percentile'`: Muestra el porcentaje acumularo de obsevaciones en cada bin
* `'cumulative`: Muestra el número acumulado de observaciones en cada bin.

```python
# Matriz de hostogramas en seaborn
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
sns.histplot(data=tips,x='tip',
             bins=15,hue='sex',
             stat='probability',
             ax=axes[0,0])

sns.histplot(data=tips,x='tip',
             bins=15,hue='sex',
             cumulative=True,
             stat='probability',
             ax=axes[0,1])

sns.histplot(data=tips,x='total_bill',
             bins=15,hue='sex',
             stat='percent',
             ax=axes[1,0])

sns.histplot(data=tips,x='total_bill',
             bins=15,hue='sex',
             cumulative=True,
             stat='percent',
             ax=axes[1,1])

plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

La función `sns.kdeplot` en Seaborn se utiliza para visualizar estimaciones de densidad del kernel (KDE), que proporcionan una representación suavizada de la distribución de una variable continua. Este método es especialmente útil para comprender la forma y la estructura de la distribución, ya que suaviza la irregularidad inherente a los histogramas. Su sintaxis básica es `sns.kdeplot(data['variable'], shad, color)` en donde:

* `data['variable']`: La variable continua que se desea visualizar.
* `shade`: Un booleano que controla si el área bajo la curva KDE se sombrea.
* `color`: Define el color de la curva KDE.

```python
# Matriz de gráficos kde
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
sns.kdeplot(data=tips,x='tip',
            hue='sex',
            ax=axes[0,0])
sns.kdeplot(data=tips,x='tip',
            hue='sex',cumulative=True,
            ax=axes[0,1])
sns.kdeplot(data=tips,x='tip',
            hue='sex',shade=True,
            ax=axes[1,0])
sns.kdeplot(data=tips,x='tip',
            hue='sex',cumulative=True,
            shade=True,
            ax=axes[1,1])
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

La función `sns.ecdfplot` en Seaborn se utiliza para visualizar funciones de distribución acumulativa empírica (ECDF), que muestran cómo se acumulan los valores en una distribución. La ECDF es una herramienta útil para entender la distribución de datos y la probabilidad asociada a diferentes valores. Su sintaxis básica es `sns.ecdfplot(data['variable'], stat, complementary, color)`. En donde:

* `data['variable']`: La variable continua que se desea visualizar.
* `stat`: Controla el tipo de ECDF que se visualiza. Puede ser 'proportion', 'percentile' o 'kde'.
* `complementary`: Un booleano que indica si mostrar la ECDF complementaria.
* `color`: Define el color de la ECDF.

```python
# Función acumulativa
plt.figure(figsize=(6,3))
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

#**3. Plots para variables categóricas**

---
Plots para variables categóricas

---

La visualización de datos categóricos es esencial para comprender la distribución y las relaciones entre diferentes grupos en un conjunto de datos. Seaborn, una biblioteca de visualización basada en Matplotlib, proporciona herramientas poderosas y atractivas para explorar variables categóricas de manera efectiva.

```python
import seaborn as sns
sns.load_dataset('tips')
tips.head()
```

La función `sns.countplot` en Seaborn se utiliza para visualizar la distribución de frecuencias de una variable categórica. Es especialmente útil para contar la cantidad de observaciones en cada categoría y obtener una visión rápida de la proporción de cada clase en un conjunto de datos. Su sintaxis básica es `sns.countplot(x='variable', data, hue, palette, order, orient)`, en donde:

* `x`: La variable categórica que se desea visualizar en el eje x.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categórica, creando barras agrupadas o apiladas según los niveles de esta variable.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `order`: Opcional. Especifica el orden en el que se deben mostrar las categorías en el eje x.
* `orient`: Opcional. Puede ser 'v' (vertical) o 'h' (horizontal), controlando la orientación de las barras.



```python
# Countplot básico
plt.figure(figsize=(4,3))
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

La función `sns.stripplot` en Seaborn se utiliza para visualizar puntos individuales en una línea para cada categoría de una variable categórica. Es particularmente útil cuando se desea mostrar la distribución de datos y la densidad de puntos en un conjunto de categorías. Su sintaxis básica es `sns.stripplot(x='variable', y='value', data, hue, palette, dodge, jitter, marker)` en donde:

* `x`: La variable categórica que se desea visualizar en el eje x.
* `y`: La variable numérica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categórica, creando tiras separadas para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `dodge`: Opcional. Si es True, separa las tiras para cada nivel de la variable hue.
* `jitter`: Opcional. Controla la dispersión horizontal de los puntos para evitar superposiciones.
* `marker`: Opcional. Especifica el estilo del marcador para los puntos en el strip plot.


```python
# Strip plot básico con separación de hue
plt.figure(figsize=(4,3))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en función del sía y del sexo')
plt.show()
```

La función `sns.swarmplot` en Seaborn se utiliza para visualizar puntos individuales en una línea para cada categoría de una variable categórica. Al igual que `sns.stripplot`, es útil para mostrar la distribución de datos y la densidad de puntos en un conjunto de categorías. Sin embargo, a diferencia de `stripplot`, `swarmplot` evita que los puntos se superpongan, proporcionando una visión más clara de la distribución. Su sintaxis básica es `sns.swarmplot(x='variable', y='value', data, hue, palette, dodge, size, edgecolor)` en donde:

* `x`: La variable categórica que se desea visualizar en el eje x.
* `y`: La variable numérica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categórica, creando enjambres separados para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `dodge`: Opcional. Si es True, separa los enjambres para cada nivel de la variable hue.
* `size`: Opcional. Controla el tamaño de los puntos en el enjambre.
* `edgecolor`: Opcional. Define el color del borde de los puntos en el enjambre.


```python
# Swarmplot plot básico con separación de hue
plt.figure(figsize=(4,3))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en función del sía y del sexo')
plt.show()
```

La función `sns.boxplot` en Seaborn se utiliza para visualizar la distribución de una variable numérica en diferentes niveles de una variable categórica. Proporciona información sobre la mediana, cuartiles y posibles valores atípicos en cada categoría, lo que facilita la comparación de distribuciones. Su sintaxis básica es `sns.boxplot(x='variable_categorica', y='variable_numerica', data, hue, palette, order)`, en donde:

* `x`: La variable categórica que se desea visualizar en el eje x.
* `y`: La variable numérica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categórica, creando cajas separadas para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `order`: Opcional. Especifica el orden en el que se deben mostrar las categorías en el eje x.

```python
# Boxplot plot básico con separación de hue
plt.figure(figsize=(4,3))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en función del sía y del sexo')
plt.show()
```

```python
# Combinación de swarmplot y boxplot
plt.figure(figsize=(4,3))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,
              color='0',marker='<')
plt.ylabel('Total bill')
plt.xlabel('Dia')
plt.title('Propinas en función del sía y del sexo')
plt.show()
```

La función `sns.violinplot` en Seaborn se utiliza para visualizar la distribución de una variable numérica en diferentes niveles de una variable categórica. A diferencia de un box plot, violinplot combina la información de un box plot con una estimación de densidad del kernel (KDE), proporcionando una representación más completa de la distribución. Su sintaxis básica es `sns.violinplot(x='variable_categorica', y='variable_numerica', data, hue, palette, split, inner, bw) en donde:

* `x`: La variable categórica que se desea visualizar en el eje x.
* `y`: La variable numérica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categórica, creando violines separados para cada nivel.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `split`: Opcional. Si es True, divide los violines cuando hay una variable categórica en el eje x.
* `inner`: Opcional. Controla la representación interna del violin, como 'box' para mostrar un box plot en su interior.
* `bw`: Opcional. Controla la anchura del kernel en la estimación de densidad.

```python
# Combinación de violinplots, boxplots y swarmplot
figure, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,4))
sns.violinplot(data=tips,x='day',y='total_bill',
               hue='sex',dodge=True,ax=axes[0])
sns.swarmplot(data=tips,x='day',y='total_bill',
              hue='sex',dodge=True,color='0',marker='<',
              ax=axes[0])
sns.violinplot(data=tips,x='day',y='total_bill',
               hue='sex',dodge=False,split=False,
               ax=axes[1])

plt.subplots_adjust(wspace=0.4)
axes[0].set_ylabel('Total bill')
axes[1].set_ylabel('Total bill')
axes[0].set_xlabel('Day')
axes[1].set_xlabel('Day')
axes[0].set_title('Tips as a function of the day and sex')
axes[1].set_title('Tips as a function of the day and sex')
plt.show()
```

La función `sns.catplot` en Seaborn es una herramienta versátil para visualizar variables categóricas y numéricas en diferentes tipos de plots categóricos. catplot permite explorar patrones y relaciones en datos complejos mediante la creación de subplots basados en la estructura de un conjunto de datos. Su sintaxis básica es `sns.catplot(x='variable_categorica', y='variable_numerica', data, kind, hue, palette, col, row), en donde:

* `x`: La variable categórica que se desea visualizar en el eje x.
* `y`: La variable numérica que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `kind`: Tipo de plot categórico a utilizar ('strip', 'swarm', 'box', 'violin', etc.).
* `hue`: Opcional. Permite segmentar los datos por una segunda variable categórica.
* `palette`: Opcional. Define la paleta de colores a utilizar en el plot.
* `col y row`: Opcionales. Permiten segmentar el plot en columnas y filas según variables categóricas adicionales.

```python
# Catplot básico
sns.catplot(data=tips,x='day',y='total_bill',
            hue='sex',dodge=True,kind='box',
            col='time')
plt.show()
```

# **4. Plots para relaciones entre variables**

---
Plots para relaciones entre variables

---

La visualización de relaciones entre variables es esencial en el análisis de datos, ya que permite comprender patrones, tendencias y posibles correlaciones. Seaborn ofrece herramientas poderosas para crear gráficos que destacan las relaciones entre diferentes variables. A continuación, se presenta una introducción a algunos de los plots más utilizados para explorar relaciones en datos.

```python
tips_data=sns.load_dataset('tips')
```

La función `sns.scatterplot` en Seaborn se utiliza para crear scatter plots, que son gráficos que muestran la relación entre dos variables numéricas mediante la colocación de puntos en un plano cartesiano. Este tipo de visualización es valioso para identificar patrones, tendencias y posibles correlaciones entre variables. Acepta parámetros como:

* `x`: Especifica la variable que se mapeará que en el eje x.
* `y`: Especifica la variable que se mapeará en el eje y.
* `data`: DataFrame que contiene los datos.
* `hue`: Permite segmentar los datos por una tercera variable categórica, coloreando los puntos.
* `style`: Permite segmentar los datos por una cuarta variable categórica, usando diferentes estilos de marcador.
* `size`: Permite segmentar los datos por una quinta variable numérica, ajustando el tamaño de los marcadores.

```python
# Scatter plot con múltiples divisiones
sns.scatterplot(data=tips_data,x='total_bill',y='tip',
                hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.15,0.5))
plt.grid(True)
plt.show()
```

```python
# Scatter plot con múltiples divisiones y markers definidos
plt.figure(figsize=(8,8))
markers = {'Lunch':'D','Dinner':'s'}
sns.scatterplot(data=tips_data,x='total_bill',y='tip',
                hue='day',style='time',size='size',
                markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.15,0.5))
plt.grid(True)
plt.show()
```

La función `sns.relplot` en Seaborn se utiliza para crear gráficos que muestran la relación entre dos variables numéricas, permitiendo segmentar y agregar dimensiones adicionales para explorar patrones más complejos. Este método es especialmente útil para visualizar la relación entre variables en conjuntos de datos más grandes. Su sintaxis básica es `sns.relplot(x='variable1', y='variable2', data, hue, style, size, col, row, kind)`, en donde:

* `x`: Variable que se mapea en el eje x.
* `y`: Variable que se mapea en el eje y.
* `data`: El DataFrame que contiene los datos.
* `hue, style, size, col, row`: Opcionales. Permiten segmentar y agregar dimensiones adicionales a los datos.
* `kind`: Opcional. Especifica el tipo de gráfico (por defecto, 'scatter, aunque puede tomar otros valores como 'line', 'linplot', 'reg', 'lm').

```python
# Relplot para mostrar múltiples divisiones de los datos
plt.figure(figsize=(8,8))
markers = {'Lunch':'D','Dinner':'s'}
sns.relplot(data=tips_data,x='total_bill',y='tip',
            hue='day',style='time',size='size',
            markers=markers,
            kind='scatter',col='time')

plt.show()
```

La función `sns.jointplot` en Seaborn se utiliza para crear gráficos que muestran la relación conjunta entre dos variables y las distribuciones marginales de cada variable. Este método es útil para analizar la distribución bivariada y examinar patrones de correlación al mísmo tiempo. Su sintaxis básica es `sns.jointplot(x='variable1', y='variable2', data, kind)` , en donde:

* `x` : Especifica la variable que se mapeará en el eje x.
* `y` : Especifica la variable que se mapeará en el eje y.
* `data` : El DataFrame que contiene los datos.
* `kind` : Especifica el tipo de gráfico a crear. Por defecto, es `'scatter'`, pero puede tomar los valores `'hex'`, `'kde'`, `'reg'` , `'resid'`, `'hist'`.
* `hue` : Determina la variable con base a la cual se dividen los datos.

```python
# Jointplot de dos variables de tipo scatter
sns.jointplot(data=tips_data, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
# Jointplot de dos variables de tipo hex
sns.jointplot(data=tips_data, x='total_bill', y='tip',
              kind='hex', marginal_ticks=True)
plt.show()
```

La función `sns.pairplot` en Seaborn se utiliza para crear una matriz de gráficos que muestra las relaciones bivariadas entre múltiples variables en un conjunto de datos. Es una herramienta poderosa para explorar patrones y correlaciones en datos multidimensionales. Su sintaxis básica es `sns.pairplot(data, hue, palette, kind, diag_kind)`, en donde:

* `data`: El DataFrame que contiene los datos.
* `hue`: Opcional. Permite segmentar los datos por una variable categórica, coloreando los puntos.
* `palette`: Opcional. Define la paleta de colores a utilizar cuando se utiliza la opción hue.
* `kind`: Opcional. Especifica el tipo de gráfico para las relaciones bivariadas (por defecto, 'scatter').
* `diag_kind`: Opcional. Especifica el tipo de gráfico para las distribuciones marginales (por defecto, 'auto').

```python
# Pairplot simple
sns.pairplot(data=tips_data)
plt.show()
```

# **5. Heatmaps**

La función `sns.heatmap` en Seaborn se utiliza para crear mapas de calor, que son visualizaciones efectivas para representar matrices de datos bidimensionales. Es especialmente útil para resaltar patrones y tendencias en datos tabulares. Su sintaxis básica es `sns.heatmap(data, annot, cmap, linewidths, cbar)`, en donde:

* `data` : La matriz bidimensional o el DataFrame que se desea visualizar.
* `annot` : Si es `True`, muestra los valores dentro de cada celda.
* `cmap` : Especifica la paleta de colores a utilizar.
* `linewidths` : Ancho de las líneas divisorias entre celdas.
* `cbar` : Si es `True`, muestra una barra de color que indica la convención de color empleada.

```python
# Heatmap de la correlación entre variables
cor = tips_data.corr()
sns.heatmap(cor,annot=True, cmap='coolwarm',linewidths=5,
            linecolor='black', vmin=0.4, vmax=1,
            cbar=True)
plt.show()
```
