Matplotlib es una biblioteca de visualización de datos en Python que proporciona una amplia variedad de gráficos y herramientas para la creación de visualizaciones estáticas, interactivas y animadas.

Algunas características clave de Matplotlib incluyen:

* ***Amplia Variedad de Gráficos:*** Matplotlib ofrece soporte para una variedad de gráficos, como gráficos de líneas, barras, dispersión, histogramas, gráficos de torta, gráficos de contorno y más.

* ***Compatibilidad con NumPy:*** Matplotlib se integra bien con la biblioteca NumPy, lo que facilita la visualización de datos almacenados en arreglos NumPy.

* ***Interactividad:*** Aunque Matplotlib es conocido por visualizaciones estáticas, también se puede integrar con herramientas de interfaz de usuario para crear gráficos interactivos.

Matplotlib se ha convertido en la base de muchas otras bibliotecas de visualización de datos en Python y es ampliamente utilizado en la comunidad científica y de análisis de datos.

Matplotlib fue creada por John D. La biblioteca se originó como un proyecto personal de Hunter, pero rápidamente ganó popularidad dentro de la comunidad científica y de análisis de datos debido a su versatilidad y facilidad de uso. A lo largo de los años, Matplotlib ha experimentado varias actualizaciones y mejoras, gracias al trabajo continuo de una comunidad activa de desarrolladores.

En 2012, John D. Hunter falleció, pero su legado continuó y la comunidad de Matplotlib siguió creciendo y mejorando la biblioteca. La falta de su líder original llevó a un esfuerzo conjunto para mantener y expandir la funcionalidad de Matplotlib.

Hoy en día, Matplotlib sigue siendo una de las bibliotecas más utilizadas para la visualización de datos en Python, y su éxito ha llevado al desarrollo de otras bibliotecas basadas en ella, como Seaborn y Pandas Plotting. La capacidad de Matplotlib para crear gráficos altamente personalizables y su integración con otras bibliotecas y entornos de desarrollo hacen que sea una herramienta esencial en el ámbito de la ciencia de datos y la visualización de datos en Python.

###1.1. Scatter plots y descriptores estadísticos básicos

Para crear un scatter plot sencillo empleamos el comando `plt.plot(x, y, fmt, **kwargs)` en donde:
* `x`: Secuencia de valores en el eje x (opcional).
* `y`: Secuencia de valores para el eje y.
* `fmt`: Especifica el formato del gráfico, como el tipo de línea y el color.
* `kwargs`: Puede incluir otros argumentos opcionales para personalizar el gráfico, como etiquetas, colores, tamaños, etc.

Dentro del argumento `fmt` se especifican las características del plot en forma de un string con las especificaciones de usuario, algúnos de los comandos para formato son:

* **Color:**
  * `'b'`: Azul (blue)
  * `'g'`: Verde (green)
  * `'r'`: Rojo (red)
  * `'c'`: Cian (cyan)
  * `'m'`: Magenta
  * `'y'`: Amarillo (yellow)
  * `'k'`: Negro (black)
  * `'w'`: Blanco (white)
* **Tipo de línea:**
  * `'-'`: Línea sólida
  * `'--'`: Línea discontinua
  * `':'`: Línea punteada
  * `'-.'`: Línea-punto
* **Marcadores:**
  * `'.'`: Punto
  * `'o'`: Círculo
  * `'s'`: Cuadrado
  * `'D'`: Diamante
  * `'^'`: Triángulo hacia arriba
  * `'v'`: Triángulo hacia abajo
  * `'<'`: Triángulo hacia la izquierda
  * `'>'`: Triángulo hacia la derecha




```python
# Scatter plot sencillo
x = np.linspace(-5,5,12)
y = x**3
plt.plot(x,y,'bo-.')
plt.grid(True)
plt.show()
```

```python
# Scatter plot sencillo
x = np.linspace(-5,5,12)
y = x**3
plt.plot(x,y,'gD-.')
plt.grid(True)
plt.show()
```

Otra forma de generar plots de scatter es con el comando `plt.scatter` cuya sintaxis es `plt.scatter(x, y, s, c, marker, cmap, norm, vmin, vmax, alpha, linewidths, edgecolors, label)` en donde:

* `x`: Secuencia de valores para el eje x.
* `y`: Secuencia de valores para el eje y.
* `s`: Tamaño de los puntos (opcional).
* `c`: Color o secuencia de colores de los puntos (opcional).
* `marker`: Tipo de marcador (por ejemplo, `'o'` para círculos, `'^'` para triángulos) (opcional).
* `cmap`: Mapa de colores para la escala de colores (opcional).
* `norm`: Normalización de colores (opcional).
* `vmin, vmax`: Valores mínimo y máximo para la escala de colores (opcional).
alpha: Transparencia de los puntos (opcional).
* `linewidths`: Ancho de las líneas del contorno de los puntos (opcional).
* `edgecolors`: Color del contorno de los puntos (opcional).
* `label`: Etiqueta para la leyenda (opcional).

```python
# Scatter plot sencillo con scatter
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)
plt.scatter(x, y,
            color='blue',
            marker='o',
            label='Puntos de datos')
plt.legend()
plt.grid(True)
plt.show()
```

El comando `hist` se utiliza en Matplotlib para crear histogramas, los cuales son gráficos que representan la distribución de frecuencias de un conjunto de datos. Estos gráficos dividen el rango de valores en intervalos (bins) y muestran cuántas observaciones caen en cada intervalo. La sintaxis del comando `hist` es `plt.hist(x, bins, range, density, cumulative, color, edgecolor, linewidth, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, label)` en donde:

* `x`: Conjunto de datos que se va a representar en el histograma.
* `bins`: Especifica el número de contenedores o una secuencia de límites de los contenedores.
* `range`: Tupla que define el rango de los datos (opcional).
* `density`: Si es True, la altura de cada bin se normaliza para formar una densidad de probabilidad (opcional).
* `cumulative`: Si es True, se muestra un histograma acumulativo (opcional).
* `color`: Color de las barras del histograma (opcional).
* `edgecolor`: Color del borde de las barras (opcional).
* `linewidth`: Ancho del borde de las barras (opcional).
* `histtype`: Tipo de histograma ('bar', 'barstacked', 'step', 'stepfilled') (opcional).
* `align`: Controla la alineación de los bins ('left', 'mid', 'right') (opcional).
* `orientation`: Orientación del histograma ('vertical' u 'horizontal') (opcional).
* `rwidth`: Ancho relativo de las barras como un porcentaje del ancho del bin (opcional).
* `log`: Si es True, el eje y se representa en escala logarítmica (opcional).
* `label`: Etiqueta para la leyenda (opcional).

```python
data = np.random.randn(1000)
plt.hist(
    data,
    bins = 30,
    color = 'blue',
    edgecolor = 'black',
    alpha = 0.7,
    label = 'Datos aleatorios'
)
plt.grid(True)
plt.show()
```

El comando `pie` se utiliza en Matplotlib para crear gráficos de tarta (o gráficos circulares), que son ideales para representar proporciones o porcentajes de un conjunto de datos en relación con el total. La sintaxis del comando `pie` es `plt.pie(x, explode, labels, colors, autopct, pctdistance, shadow, startangle, radius, counterclock, wedgeprops, textprops, center, frame, rotatelabels, normalize)` en donde:

* `x : Conjunto de datos que se va a representar en el gráfico de tarta.
* `explode : Lista o array que especifica la separación de una o más porciones (opcional).
* `labels`: Lista de etiquetas para las porciones (opcional).
* `colors`: Lista de colores para las porciones (opcional).
* `autopct`: Especifica el formato de la etiqueta automática de porcentaje (opcional).
* `pctdistance`: Distancia del porcentaje desde el centro del círculo (opcional).
* `shadow`: Si es True, se añade sombra al gráfico (opcional).
* `startangle`: Ángulo de inicio para el primer segmento (opcional).
* `radius`: Radio del círculo (opcional).
* `counterclock`: Si es True, el gráfico se dibuja en sentido antihorario (opcional).
* `wedgeprops`: Propiedades de las porciones (opcional).
* `textprops`: Propiedades del texto (opcional).
* `center`: Coordenadas del centro del círculo (opcional).
* `frame`: Si es True, se dibuja un marco alrededor del gráfico (opcional).
* `rotatelabels`: Si es True, se rotan las etiquetas (opcional).
* `normalize`: Si es True, los valores de x se normalizan para sumar 100% (opcional).

```python
# Pie plot simple
labels = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']
sizes = [33, 20,17,40]
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=0,
    colors=['orange', 'green', 'blue','red'],
    shadow=True
    )
plt.show()
```

El comando `boxplot` en Matplotlib se utiliza para crear diagramas de caja y bigotes, también conocidos como boxplots. Estos gráficos son útiles para visualizar la distribución y estadísticas resumidas de un conjunto de datos, como la mediana, cuartiles y posibles valores atípicos. Su sintaxis básica es `plt.boxplot(x, notch, sym, vert, whis, positions, widths, patch_artist, bootstrap, usermedians, conf_intervals, meanline, showmeans, showcaps, showbox, showfliers, boxprops, whiskerprops, capprops, flierprops, medianprops, meanprops, manage_ticks)` en dónde:

* `x`: Conjunto de datos que se va a representar en el boxplot.
* `notch`: Si es True, se dibuja una muesca en la caja para indicar la incertidumbre acerca de la mediana (opcional).
* `sym`: Símbolos para representar los posibles valores atípicos (opcional).
* `vert`: Si es True, el boxplot es vertical; si es False, es horizontal (opcional).
* `whis`: Proporción del rango intercuartílico para extender los bigotes (opcional).
* `positions`: Posiciones de los boxplots en el eje x (opcional).
* `widths`: Ancho de los boxplots (opcional).
* `patch_artist`: Si es True, los componentes del boxplot se representan como objetos de parche (opcional).
* `bootstrap`: Número de arranques para calcular intervalos de confianza (opcional).
* `usermedians`: Posiciones personalizadas de las medianas (opcional).
* `conf_intervals`: Intervalos de confianza para las medianas (opcional).
* `meanline`: Si es True, se dibuja una línea que representa la media (opcional).
* `showmeans`: Si es True, se muestran los valores medios (opcional).
* `showcaps, showbox, showfliers`: Controlan la visibilidad de las tapas, la caja y los valores atípicos (opcional).
* `boxprops, whiskerprops, capprops, flierprops, medianprops, meanprops`: Diccionarios con propiedades de los componentes del boxplot (opcional).
* `manage_ticks`: Si es True, Matplotlib gestiona las marcas de los ejes automáticamente (opcional).

```python
# Boxplot básico
data = np.random.randn(100)
plt.boxplot(
    data,
    vert=True,
    patch_artist=True
    )
plt.grid(True)
plt.show()
```

```python
# Múltiples box plots
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data,
            vert=True,
            patch_artist=True,
            widths=0.5,
            labels=['A', 'B', 'C'],
            showmeans=True,
            meanline=True,
            showfliers=True)

# Personalizar el gráfico
plt.title('Boxplot de Datos')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

### 1.2. Subplots

---
Subplots

---

El comando `subplot` en Matplotlib se utiliza para crear una disposición de subgráficos en una figura. Permite dividir la ventana de visualización en una cuadrícula y colocar gráficos individuales en ubicaciones específicas dentro de esa cuadrícula. Su sintaxis básica es `plt.subplot(nrows, ncols, index, **kwargs)` en donde:

* `nrows`: Número de filas en la cuadrícula.
* `ncols`: Número de columnas en la cuadrícula.
* `index`: Índice del subgráfico actual (comienza desde 1 y va de izquierda a derecha y de arriba a abajo).
* `**kwargs`: Puedes proporcionar argumentos adicionales que son pasados a la función add_subplot.


```python
# Graficar dos plots en una mísma figura
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# -------Crear una figura con subgráficos
# Subgráfico uno
num_filas = 1
cantidad_columnas = 1
indice_figura = 1

plt.subplot(num_filas, cantidad_columnas, indice_figura)
plt.plot(x, y1, label='sin(x)')
plt.title('Subgráfico 1')
plt.grid(True)
plt.legend()

# Subgráfico dos
plt.subplot(num_filas, cantidad_columnas, indice_figura)
plt.plot(x, y2, label='cos(x)')
plt.title('Subgráfico 2')
plt.grid(True)
plt.legend()

# Ajustar el diseño y mostrar la figura
plt.tight_layout()
plt.show()
```

# **2. Métodos orientados a objetos**

---
Métodos orientados a objetos

---

Los métodos orientados a objetos en Matplotlib proporcionan una forma más flexible y poderosa de crear y personalizar múltiples gráficas en una figura. En lugar de utilizar funciones como `plt.plot()` o `plt.subplot()`, se trabaja directamente con objetos de gráfica y ejes.
Para comenzar, se crea una figura y un conjunto de ejes utilizando el método `subplots()`, en donde `fig` es la figura global que contiene todos los elementos del gráfico, y `ax` es un conjunto de ejes dentro de esa figura.

### 3.1. Datos por pares

---
Datos por pares

---
Los conjuntos de visualizaciónes por pares corresponden a gráficos que permiten visualizar cierta relación entre dos colecciones de datos, por ejemplo, la relación entre una variable y otra.

```python
# Gráfico scatter simple
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Creción de una figura y unos ejes
fig, ax = plt.subplots()
ax.plot(x , y, label='sin(x)')
ax.plot(x + 1, (1.5)*y, 'ro-.', label='Scatter sin(x)')
ax.plot(x + 2, (2)*y, 'go-.', label='Scatter sin(x) 2')

# Personalizar el gráfico
ax.set_title('Gráfico de Sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.grid(True)
ax.legend()
plt.show()
```

```python
# Scatter plots simples con unión de ejes
x = np.linspace(-5,5,100)
y = x**3

dimensiones = (8,4)
fig, ax1 = plt.subplots(figsize=dimensiones)
ax1.plot(5*x,y,'blue')

# Creación de un eje gemelo
ax2 = ax1.twinx()
ax2.plot(y,x,'orange')

# Personalización del gráfico
ax1.set_xlabel('x')
ax2.set_xlabel('x')
ax1.set_ylabel('x^3')
ax2.set_ylabel('x^-3')
ax1.set_title('Gráfico con eje y gemelo')

# fig.set_size_inches(5,3) tamaño de la figura en pulgadas
fig.set_dpi(100) # resolución de la figura en puntos por pulgada (DPI)
```

Para crear subplots en donde se agreguen manualmente las especificaciónes a cerca de la posición de un subgráfico y pueda ser este usado, por ejemplo, como una plot de detalle embedido en otro usamos `fig_add_axes` comando el cual permite especificar exáctamente donde se quiere colocar los ejes dentro de la figura deacuerdo a `fig_add_axes([left, bottom, width, height])`.

```python
# Subplot anidado
x = np.linspace(-5,5,100)
y = np.sin(x + 18)

# Creación de un gráfico dentro de otro
dimensiones = (8,4)
fig = plt.figure(figsize=dimensiones)
ax_1 = fig.add_axes([0.1,0.1,1,1])
ax_2 = fig.add_axes([0.13,0.15,0.3,0.5])
ax_1.plot(x,y,'b')
ax_2.plot(y,x,'r--')
ax_2.set_facecolor('white')
ax_1.set_title('Ejemplo de plot anidado')
ax_1.grid(True)
ax_2.grid(True)
plt.show()
```

```python
# Gráfico con múltiples subplots indicados por ejes
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x**2)
y2 = np.cos(x**2)

# Creación de múltiples subplots
filas = 1
columnas = 2
dimensiones = (8,4)
fig, ax = plt.subplots(filas, columnas , figsize=dimensiones )

ax[0].plot(x , y1, label='sin(x)')
ax[0].plot(x + 1, (1.5)*y1, 'ro-.', label='Scatter sin(x^2)')
ax[0].plot(x + 2, (2)*y1, 'go-.', label='Scatter sin(x^2) 2')

ax[1].plot(x , y2, label='sin(x)')
ax[1].plot(x + 1, (1.5)*y2, 'rD-.', label='Scatter cos(x^2)')
ax[1].plot(x + 2, (2)*y2, 'g<-.', label='Scatter cos(x^2) 2')

# Personalizar el gráfico
ax[0].set_title('Gráfico de Sin(x^2)')
ax[1].set_title('Gráfico de Cos(x^2)')
ax[0].set_xlabel('x')
ax[1].set_xlabel('x')
ax[0].set_ylabel('sin(x)')
ax[1].set_ylabel('cos(x)')
ax[0].grid(True)
ax[0].legend()
ax[1].grid(True)
ax[1].legend()
plt.show()
```

```python
# Gráfico matrices de gráficos
x = np.linspace(0,10, 100)

# Creación de múltiples subplots
filas = 2
columnas = 2
dimensiones = (10,5)
fig, ax = plt.subplots(filas, columnas , figsize=dimensiones )

ax[0,0].plot(x,np.sin(x),'b*-', label='sin(x)')
ax[0,1].plot(x,np.cos(x),'g*-',label='cos(x)')
ax[1,0].plot(x,np.tan(x),'r*-.',label='tan(x)')
ax[1,1].plot(x, np.cos(x),'orange',label='cos 1')
ax[1,1].plot(x + 3, np.cos(x),'blue',label='cos 2')

titulos = [['Gráfico seno', 'Gráfico coseno'], ['Gráfico tangente', 'Gráfico de cosenos']]
xlabels = [['x','x'], ['x','x']]
ylabels = [['sin(x)','cos(x)'],['tan(x)','cosenos']]

for i in range(2):
  for j in range(2):
    ax[i,j].set_title(titulos[i][j])
    ax[i,j].set_xlabel(xlabels[i][j])
    ax[i,j].set_ylabel(ylabels[i][j])
    ax[i,j].grid(True)
    ax[i,j].legend(loc='upper right', bbox_to_anchor=(1,1))


fig.tight_layout()
plt.show()
```

### 3.2. Barplots

---
Barplots

---

Los gráficos de barras (bar plots) son visualizaciones efectivas para representar datos discretos o categóricos mediante barras rectangulares. Cada barra representa una categoría y la altura de la barra corresponde a la magnitud de la variable que se está visualizando. Matplotlib proporciona funciones para crear fácilmente gráficos de barras, permitiendo una personalización detallada. Su sintaxis básica es `plt.bar(x, height, width, bottom, align, color, edgecolor, linewidth, tick_label, label, log, **kwargs)` en donde:

* `x`: Coordenadas de las barras. Pueden ser secuencias de números o nombres de categorías.
* `height`: Altura de las barras, indicando los valores que representan.
* `width`: Ancho de las barras. Por defecto es 0.8.
* `bottom`: Coordenadas en el eje y donde comienzan las barras. Por defecto es None, lo que significa que comienzan en 0.
* `align`: Controla la alineación de las barras respecto a las coordenadas x. Puede ser 'center' (centro), 'edge' (borde), o 'edge_center' (centro del borde). Por defecto es 'center'.
color: Color de las barras.
* `edgecolor`: Color del borde de las barras.
* `linewidth`: Ancho del borde de las barras.
* `tick_label`: Etiquetas para las marcas del eje x.
label: Etiqueta para la leyenda.
* `log`: Si es True, se utiliza escala logarítmica en el eje y.
* `**kwargs`: Otros argumentos opcionales para personalizar la apariencia de las barras.

```python
# Gráfico de barras apiladas
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 1, 5]
values2 = [2, 5, 3, 8]

# Crear un gráfico de barras apiladas
plt.figure(figsize=(8,4))
plt.bar(categories, values1, color='blue', label='Grupo 1')
plt.bar(categories, values2,
        bottom=values1,
        color='orange', label='Grupo 2')

# Personalizar el gráfico
plt.title('Gráfico de Barras Apiladas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()

plt.grid(True)
plt.show()
```

```python
# Gráfico de barras agrupadas
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 1, 5]
values2 = [2, 5, 3, 8]

# Coordenadas de las barras agrupadas
x = np.arange(len(categories))

# Crear un gráfico de barras agrupadas
plt.figure(figsize=(8,4))
width = 0.35
plt.bar(x - width/2, values1, width, label='Grupo 1', color='blue')
plt.bar(x + width/2, values2, width, label='Grupo 2', color='orange')

# Personalizar el gráfico
plt.title('Gráfico de Barras Agrupadas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.xticks(x, categories)
plt.legend()

# Mostrar el gráfico
plt.show()
```
