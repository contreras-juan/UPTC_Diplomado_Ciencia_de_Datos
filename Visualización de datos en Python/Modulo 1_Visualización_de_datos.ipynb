###1.1. Scatter plots y descriptores estadsticos bsicos

Para crear un scatter plot sencillo empleamos el comando `plt.plot(x, y, fmt, **kwargs)` en donde:
* `x`: Secuencia de valores en el eje x (opcional).
* `y`: Secuencia de valores para el eje y.
* `fmt`: Especifica el formato del grfico, como el tipo de lnea y el color.
* `kwargs`: Puede incluir otros argumentos opcionales para personalizar el grfico, como etiquetas, colores, tamaos, etc.

Dentro del argumento `fmt` se especifican las caractersticas del plot en forma de un string con las especificaciones de usuario, algnos de los comandos para formato son:

* **Color:**
  * `'b'`: Azul (blue)
  * `'g'`: Verde (green)
  * `'r'`: Rojo (red)
  * `'c'`: Cian (cyan)
  * `'m'`: Magenta
  * `'y'`: Amarillo (yellow)
  * `'k'`: Negro (black)
  * `'w'`: Blanco (white)
* **Tipo de lnea:**
  * `'-'`: Lnea slida
  * `'--'`: Lnea discontinua
  * `':'`: Lnea punteada
  * `'-.'`: Lnea-punto
* **Marcadores:**
  * `'.'`: Punto
  * `'o'`: Crculo
  * `'s'`: Cuadrado
  * `'D'`: Diamante
  * `'^'`: Tringulo hacia arriba
  * `'v'`: Tringulo hacia abajo
  * `'<'`: Tringulo hacia la izquierda
  * `'>'`: Tringulo hacia la derecha




```python
# Scatter plot sencillo
x = np.linspace(-5,5,12)
y = x**3
plt.plot(x,y,'bo-.')
plt.grid(True)
plt.show()
```

```python
# Scatter plot sencillo
x = np.linspace(-5,5,12)
y = x**3
plt.plot(x,y,'gD-.')
plt.grid(True)
plt.show()
```

Otra forma de generar plots de scatter es con el comando `plt.scatter` cuya sintaxis es `plt.scatter(x, y, s, c, marker, cmap, norm, vmin, vmax, alpha, linewidths, edgecolors, label)` en donde:

* `x`: Secuencia de valores para el eje x.
* `y`: Secuencia de valores para el eje y.
* `s`: Tamao de los puntos (opcional).
* `c`: Color o secuencia de colores de los puntos (opcional).
* `marker`: Tipo de marcador (por ejemplo, `'o'` para crculos, `'^'` para tringulos) (opcional).
* `cmap`: Mapa de colores para la escala de colores (opcional).
* `norm`: Normalizacin de colores (opcional).
* `vmin, vmax`: Valores mnimo y mximo para la escala de colores (opcional).
alpha: Transparencia de los puntos (opcional).
* `linewidths`: Ancho de las lneas del contorno de los puntos (opcional).
* `edgecolors`: Color del contorno de los puntos (opcional).
* `label`: Etiqueta para la leyenda (opcional).

```python
# Scatter plot sencillo con scatter
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)
plt.scatter(x, y,
            color='blue',
            marker='o',
            label='Puntos de datos')
plt.legend()
plt.grid(True)
plt.show()
```

El comando `hist` se utiliza en Matplotlib para crear histogramas, los cuales son grficos que representan la distribucin de frecuencias de un conjunto de datos. Estos grficos dividen el rango de valores en intervalos (bins) y muestran cuntas observaciones caen en cada intervalo. La sintaxis del comando `hist` es `plt.hist(x, bins, range, density, cumulative, color, edgecolor, linewidth, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, label)` en donde:

* `x`: Conjunto de datos que se va a representar en el histograma.
* `bins`: Especifica el nmero de contenedores o una secuencia de lmites de los contenedores.
* `range`: Tupla que define el rango de los datos (opcional).
* `density`: Si es True, la altura de cada bin se normaliza para formar una densidad de probabilidad (opcional).
* `cumulative`: Si es True, se muestra un histograma acumulativo (opcional).
* `color`: Color de las barras del histograma (opcional).
* `edgecolor`: Color del borde de las barras (opcional).
* `linewidth`: Ancho del borde de las barras (opcional).
* `histtype`: Tipo de histograma ('bar', 'barstacked', 'step', 'stepfilled') (opcional).
* `align`: Controla la alineacin de los bins ('left', 'mid', 'right') (opcional).
* `orientation`: Orientacin del histograma ('vertical' u 'horizontal') (opcional).
* `rwidth`: Ancho relativo de las barras como un porcentaje del ancho del bin (opcional).
* `log`: Si es True, el eje y se representa en escala logartmica (opcional).
* `label`: Etiqueta para la leyenda (opcional).

```python
data = np.random.randn(1000)
plt.hist(
    data,
    bins = 30,
    color = 'blue',
    edgecolor = 'black',
    alpha = 0.7,
    label = 'Datos aleatorios'
)
plt.grid(True)
plt.show()
```

El comando `pie` se utiliza en Matplotlib para crear grficos de tarta (o grficos circulares), que son ideales para representar proporciones o porcentajes de un conjunto de datos en relacin con el total. La sintaxis del comando `pie` es `plt.pie(x, explode, labels, colors, autopct, pctdistance, shadow, startangle, radius, counterclock, wedgeprops, textprops, center, frame, rotatelabels, normalize)` en donde:

* `x : Conjunto de datos que se va a representar en el grfico de tarta.
* `explode : Lista o array que especifica la separacin de una o ms porciones (opcional).
* `labels`: Lista de etiquetas para las porciones (opcional).
* `colors`: Lista de colores para las porciones (opcional).
* `autopct`: Especifica el formato de la etiqueta automtica de porcentaje (opcional).
* `pctdistance`: Distancia del porcentaje desde el centro del crculo (opcional).
* `shadow`: Si es True, se aade sombra al grfico (opcional).
* `startangle`: ngulo de inicio para el primer segmento (opcional).
* `radius`: Radio del crculo (opcional).
* `counterclock`: Si es True, el grfico se dibuja en sentido antihorario (opcional).
* `wedgeprops`: Propiedades de las porciones (opcional).
* `textprops`: Propiedades del texto (opcional).
* `center`: Coordenadas del centro del crculo (opcional).
* `frame`: Si es True, se dibuja un marco alrededor del grfico (opcional).
* `rotatelabels`: Si es True, se rotan las etiquetas (opcional).
* `normalize`: Si es True, los valores de x se normalizan para sumar 100% (opcional).

```python
# Pie plot simple
labels = ['Categora A', 'Categora B', 'Categora C', 'Categora D']
sizes = [33, 20,17,40]
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=0,
    colors=['orange', 'green', 'blue','red'],
    shadow=True
    )
plt.show()
```

El comando `boxplot` en Matplotlib se utiliza para crear diagramas de caja y bigotes, tambin conocidos como boxplots. Estos grficos son tiles para visualizar la distribucin y estadsticas resumidas de un conjunto de datos, como la mediana, cuartiles y posibles valores atpicos. Su sintaxis bsica es `plt.boxplot(x, notch, sym, vert, whis, positions, widths, patch_artist, bootstrap, usermedians, conf_intervals, meanline, showmeans, showcaps, showbox, showfliers, boxprops, whiskerprops, capprops, flierprops, medianprops, meanprops, manage_ticks)` en dnde:

* `x`: Conjunto de datos que se va a representar en el boxplot.
* `notch`: Si es True, se dibuja una muesca en la caja para indicar la incertidumbre acerca de la mediana (opcional).
* `sym`: Smbolos para representar los posibles valores atpicos (opcional).
* `vert`: Si es True, el boxplot es vertical; si es False, es horizontal (opcional).
* `whis`: Proporcin del rango intercuartlico para extender los bigotes (opcional).
* `positions`: Posiciones de los boxplots en el eje x (opcional).
* `widths`: Ancho de los boxplots (opcional).
* `patch_artist`: Si es True, los componentes del boxplot se representan como objetos de parche (opcional).
* `bootstrap`: Nmero de arranques para calcular intervalos de confianza (opcional).
* `usermedians`: Posiciones personalizadas de las medianas (opcional).
* `conf_intervals`: Intervalos de confianza para las medianas (opcional).
* `meanline`: Si es True, se dibuja una lnea que representa la media (opcional).
* `showmeans`: Si es True, se muestran los valores medios (opcional).
* `showcaps, showbox, showfliers`: Controlan la visibilidad de las tapas, la caja y los valores atpicos (opcional).
* `boxprops, whiskerprops, capprops, flierprops, medianprops, meanprops`: Diccionarios con propiedades de los componentes del boxplot (opcional).
* `manage_ticks`: Si es True, Matplotlib gestiona las marcas de los ejes automticamente (opcional).

```python
# Boxplot bsico
data = np.random.randn(100)
plt.boxplot(
    data,
    vert=True,
    patch_artist=True
    )
plt.grid(True)
plt.show()
```

```python
# Mltiples box plots
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data,
            vert=True,
            patch_artist=True,
            widths=0.5,
            labels=['A', 'B', 'C'],
            showmeans=True,
            meanline=True,
            showfliers=True)

# Personalizar el grfico
plt.title('Boxplot de Datos')
plt.xlabel('Categoras')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

### 1.2. Subplots

Subplots

El comando `subplot` en Matplotlib se utiliza para crear una disposicin de subgrficos en una figura. Permite dividir la ventana de visualizacin en una cuadrcula y colocar grficos individuales en ubicaciones especficas dentro de esa cuadrcula. Su sintaxis bsica es `plt.subplot(nrows, ncols, index, **kwargs)` en donde:

* `nrows`: Nmero de filas en la cuadrcula.
* `ncols`: Nmero de columnas en la cuadrcula.
* `index`: ndice del subgrfico actual (comienza desde 1 y va de izquierda a derecha y de arriba a abajo).
* `**kwargs`: Puedes proporcionar argumentos adicionales que son pasados a la funcin add_subplot.


```python
# Graficar dos plots en una msma figura
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# -------Crear una figura con subgrficos
# Subgrfico uno
num_filas = 1
cantidad_columnas = 1
indice_figura = 1

plt.subplot(num_filas, cantidad_columnas, indice_figura)
plt.plot(x, y1, label='sin(x)')
plt.title('Subgrfico 1')
plt.grid(True)
plt.legend()

# Subgrfico dos
plt.subplot(num_filas, cantidad_columnas, indice_figura)
plt.plot(x, y2, label='cos(x)')
plt.title('Subgrfico 2')
plt.grid(True)
plt.legend()

# Ajustar el diseo y mostrar la figura
plt.tight_layout()
plt.show()
```

# **2. Mtodos orientados a objetos**

---
Mtodos orientados a objetos

---

Los mtodos orientados a objetos en Matplotlib proporcionan una forma ms flexible y poderosa de crear y personalizar mltiples grficas en una figura. En lugar de utilizar funciones como `plt.plot()` o `plt.subplot()`, se trabaja directamente con objetos de grfica y ejes.
Para comenzar, se crea una figura y un conjunto de ejes utilizando el mtodo `subplots()`, en donde `fig` es la figura global que contiene todos los elementos del grfico, y `ax` es un conjunto de ejes dentro de esa figura.

### 3.1. Datos por pares

---
Datos por pares

---
Los conjuntos de visualizacines por pares corresponden a grficos que permiten visualizar cierta relacin entre dos colecciones de datos, por ejemplo, la relacin entre una variable y otra.

```python
# Grfico scatter simple
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Crecin de una figura y unos ejes
fig, ax = plt.subplots()
ax.plot(x , y, label='sin(x)')
ax.plot(x + 1, (1.5)*y, 'ro-.', label='Scatter sin(x)')
ax.plot(x + 2, (2)*y, 'go-.', label='Scatter sin(x) 2')

# Personalizar el grfico
ax.set_title('Grfico de Sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.grid(True)
ax.legend()
plt.show()
```

```python
# Scatter plots simples con unin de ejes
x = np.linspace(-5,5,100)
y = x**3

dimensiones = (8,4)
fig, ax1 = plt.subplots(figsize=dimensiones)
ax1.plot(5*x,y,'blue')

# Creacin de un eje gemelo
ax2 = ax1.twinx()
ax2.plot(y,x,'orange')

# Personalizacin del grfico
ax1.set_xlabel('x')
ax2.set_xlabel('x')
ax1.set_ylabel('x^3')
ax2.set_ylabel('x^-3')
ax1.set_title('Grfico con eje y gemelo')

# fig.set_size_inches(5,3) tamao de la figura en pulgadas
fig.set_dpi(100) # resolucin de la figura en puntos por pulgada (DPI)
```

Para crear subplots en donde se agreguen manualmente las especificacines a cerca de la posicin de un subgrfico y pueda ser este usado, por ejemplo, como una plot de detalle embedido en otro usamos `fig_add_axes` comando el cual permite especificar exctamente donde se quiere colocar los ejes dentro de la figura deacuerdo a `fig_add_axes([left, bottom, width, height])`.

```python
# Subplot anidado
x = np.linspace(-5,5,100)
y = np.sin(x + 18)

# Creacin de un grfico dentro de otro
dimensiones = (8,4)
fig = plt.figure(figsize=dimensiones)
ax_1 = fig.add_axes([0.1,0.1,1,1])
ax_2 = fig.add_axes([0.13,0.15,0.3,0.5])
ax_1.plot(x,y,'b')
ax_2.plot(y,x,'r--')
ax_2.set_facecolor('white')
ax_1.set_title('Ejemplo de plot anidado')
ax_1.grid(True)
ax_2.grid(True)
plt.show()
```

```python
# Grfico con mltiples subplots indicados por ejes
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x**2)
y2 = np.cos(x**2)

# Creacin de mltiples subplots
filas = 1
columnas = 2
dimensiones = (8,4)
fig, ax = plt.subplots(filas, columnas , figsize=dimensiones )

ax[0].plot(x , y1, label='sin(x)')
ax[0].plot(x + 1, (1.5)*y1, 'ro-.', label='Scatter sin(x^2)')
ax[0].plot(x + 2, (2)*y1, 'go-.', label='Scatter sin(x^2) 2')

ax[1].plot(x , y2, label='sin(x)')
ax[1].plot(x + 1, (1.5)*y2, 'rD-.', label='Scatter cos(x^2)')
ax[1].plot(x + 2, (2)*y2, 'g<-.', label='Scatter cos(x^2) 2')

# Personalizar el grfico
ax[0].set_title('Grfico de Sin(x^2)')
ax[1].set_title('Grfico de Cos(x^2)')
ax[0].set_xlabel('x')
ax[1].set_xlabel('x')
ax[0].set_ylabel('sin(x)')
ax[1].set_ylabel('cos(x)')
ax[0].grid(True)
ax[0].legend()
ax[1].grid(True)
ax[1].legend()
plt.show()
```

```python
# Grfico matrices de grficos
x = np.linspace(0,10, 100)

# Creacin de mltiples subplots
filas = 2
columnas = 2
dimensiones = (10,5)
fig, ax = plt.subplots(filas, columnas , figsize=dimensiones )

ax[0,0].plot(x,np.sin(x),'b*-', label='sin(x)')
ax[0,1].plot(x,np.cos(x),'g*-',label='cos(x)')
ax[1,0].plot(x,np.tan(x),'r*-.',label='tan(x)')
ax[1,1].plot(x, np.cos(x),'orange',label='cos 1')
ax[1,1].plot(x + 3, np.cos(x),'blue',label='cos 2')

titulos = [['Grfico seno', 'Grfico coseno'], ['Grfico tangente', 'Grfico de cosenos']]
xlabels = [['x','x'], ['x','x']]
ylabels = [['sin(x)','cos(x)'],['tan(x)','cosenos']]

for i in range(2):
  for j in range(2):
    ax[i,j].set_title(titulos[i][j])
    ax[i,j].set_xlabel(xlabels[i][j])
    ax[i,j].set_ylabel(ylabels[i][j])
    ax[i,j].grid(True)
    ax[i,j].legend(loc='upper right', bbox_to_anchor=(1,1))


fig.tight_layout()
plt.show()
```

### 3.2. Barplots

---
Barplots

---

Los grficos de barras (bar plots) son visualizaciones efectivas para representar datos discretos o categricos mediante barras rectangulares. Cada barra representa una categora y la altura de la barra corresponde a la magnitud de la variable que se est visualizando. Matplotlib proporciona funciones para crear fcilmente grficos de barras, permitiendo una personalizacin detallada. Su sintaxis bsica es `plt.bar(x, height, width, bottom, align, color, edgecolor, linewidth, tick_label, label, log, **kwargs)` en donde:

* `x`: Coordenadas de las barras. Pueden ser secuencias de nmeros o nombres de categoras.
* `height`: Altura de las barras, indicando los valores que representan.
* `width`: Ancho de las barras. Por defecto es 0.8.
* `bottom`: Coordenadas en el eje y donde comienzan las barras. Por defecto es None, lo que significa que comienzan en 0.
* `align`: Controla la alineacin de las barras respecto a las coordenadas x. Puede ser 'center' (centro), 'edge' (borde), o 'edge_center' (centro del borde). Por defecto es 'center'.
color: Color de las barras.
* `edgecolor`: Color del borde de las barras.
* `linewidth`: Ancho del borde de las barras.
* `tick_label`: Etiquetas para las marcas del eje x.
label: Etiqueta para la leyenda.
* `log`: Si es True, se utiliza escala logartmica en el eje y.
* `**kwargs`: Otros argumentos opcionales para personalizar la apariencia de las barras.

```python
# Grfico de barras agrupadas
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 1, 5]
values2 = [2, 5, 3, 8]

# Coordenadas de las barras agrupadas
x = np.arange(len(categories))

# Crear un grfico de barras agrupadas
plt.figure(figsize=(8,4))
width = 0.35
plt.bar(x - width/2, values1, width, label='Grupo 1', color='blue')
plt.bar(x + width/2, values2, width, label='Grupo 2', color='orange')

# Personalizar el grfico
plt.title('Grfico de Barras Agrupadas')
plt.xlabel('Categoras')
plt.ylabel('Valores')
plt.xticks(x, categories)
plt.legend()

# Mostrar el grfico
plt.show()
```
