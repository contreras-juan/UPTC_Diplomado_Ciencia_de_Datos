Evento

---

Un evento es cualquier subconjunto del espacio muestral. El lector interesado puede verificar, si lo desea, que el espacio muestral $\mathcal{M}$ tiene exactamente $2^{N}$ subconjuntos.

Consideremos ahora cuatro eventos (subconjuntos) especiales de $\mathcal{M}$:

1. *azul*: el subconjunto de bolas azules;
2. *gris*: el subconjunto de bolas grises.
3. *pares*: el subconjunto de bolas pares.
4. *impares*: el subconjunto de bolas impares.
Extra: *pares azules*: el subconjunto de bolas pares azules.

Prueba: Para verificar que un espacio muestral de \(N\) elementos tenga \(2^N\) subconjuntos (eventos), se puede emplear el principio de la teora de conjuntos y la combinatoria. Este fenmeno se explica a travs del concepto de las partes de un conjunto, tambin conocido como el conjunto potencia. El conjunto potencia de cualquier conjunto es el conjunto de todos sus posibles subconjuntos, incluyendo el conjunto vaco y el conjunto en s.

Para entender paso a paso por qu un espacio muestral de \(N\) elementos tiene \(2^N\) subconjuntos, podemos describirlo as:

1. **Elemento Individual**: Para cada elemento en el espacio muestral, hay dos opciones: o el elemento est presente en un subconjunto, o no lo est.

2. **Combinaciones de Elementos**: Dado que cada elemento del espacio muestral puede estar presente o no en un subconjunto, y considerando que el espacio muestral tiene \(N\) elementos, entonces para cada elemento hay 2 opciones. Esto nos da un total de \(2 x 2 x 2 x ... x 2\) (con \(N\) factores de 2), lo cual es igual a \(2^N\).

3. **Conjunto Potencia**: Matemticamente, el conjunto de todos los posibles subconjuntos de un conjunto se llama su conjunto potencia. Si el conjunto tiene \(N\) elementos, entonces el tamao de su conjunto potencia (el nmero de subconjuntos posibles) es \(2^N\).

Ejemplo:
Si tienes un espacio muestral con 3 elementos (por ejemplo, {A, B, C}), los posibles subconjuntos seran: {}, {A}, {B}, {C}, {A, B}, {A, C}, {B, C}, {A, B, C}. Aqu puedes ver que hay \(2^3 = 8\) subconjuntos, lo que coincide con nuestra explicacin.

Esta propiedad se puede verificar empricamente para espacios muestrales pequeos mediante la enumeracin de todos los posibles subconjuntos. Para espacios muestrales grandes, se puede demostrar matemticamente o mediante programacin, usando algoritmos que generen todos los posibles subconjuntos de un conjunto dado y luego contando el nmero total de subconjuntos generados para asegurar que coincide con \(2^N\).

````python
#Para nuestro espacio muestral aleatorio:

import random

# Definir el espacio muestral
espacio_muestral = ['cara', 'cruz']

# Definir el evento deseado
evento = 'cara'

# Realizar N experimentos
N = 1000
eventos_favorables = sum(random.choice(espacio_muestral) == evento for _ in range(N))

# Calcular la probabilidad del evento
probabilidad = eventos_favorables / N
print(f"Probabilidad de obtener {evento}: {probabilidad}")
````

````python
#Para nuestro espacio muestral discreto:
import numpy as np

# Definir el espacio muestral
espacio_muestral = np.array([1, 2, 3, 4, 5, 6])

# Realizar N experimentos
N = 1000
resultados = np.random.choice(espacio_muestral, size=N)

# Definir el evento deseado (resultado par)
eventos_favorables = np.sum(resultados % 2 == 0)

# Calcular la probabilidad del evento
probabilidad = eventos_favorables / N
print(f"Probabilidad de obtener un nmero par: {probabilidad}")
````

````python
#Para nuesrro espacio muestral continuo:
import numpy as np

# Generar muestras de una distribucin normal estndar
N = 1000
muestras = np.random.normal(loc=0, scale=1, size=N)

# Definir el evento deseado (muestra < 0)
eventos_favorables = np.sum(muestras < 0)

# Calcular la probabilidad del evento
probabilidad = eventos_favorables / N
print(f"Probabilidad de que una muestra sea menor que 0: {probabilidad}")
````

### <span style="color:blue">2.4. Regla aditiva de la probabilidad</span>

---
Regla aditiva de la probabilidad

---

La probabilidad de la unin de dos eventos (subconjuntos) disyuntos (que no tiene interseccin) es la suma de la probabilidad (medida) de cada uno de ellos. En smbolos, si $A$ y $B$ son eventos disyuntos de $\mathcal{M}$, entonces

$$
\text{Prob}[A\cup B] = \text{Prob}[A] + \text{Prob}[B].
$$

Por ejemplo, observe que:

Para N= 20, con Azules=5 , Rojos=7, pares = 10 \\

$$
\text{Prob}[\text{azul}\cup\text{rojo}] = 5/20+7/20 = 12/20.
$$


Lo anterior conduce a la regla aditiva general la cual dice que

$$
\text{Prob}[A\cup B] = \text{Prob}[A] + \text{Prob}[B]-\text{Prob}[A\cap B] .
$$

En el ejemplo se tiene entonces que

$$
\text{Prob}[\text{azul}\cup\text{pares}] = 5/20 + 10/20 - 3/20 = 12/20
$$

**Nota:** Un conjunto disyunto se refiere a dos o ms conjuntos que no tienen elementos en comn. Es decir, la interseccin de estos conjuntos es vaca.

En el contexto de la probabilidad, cuando se dice que dos eventos son mutuamente excluyentes o disyuntos, significa que la ocurrencia de uno de los eventos excluye automticamente la ocurrencia del otro. Por lo tanto, no pueden ocurrir al mismo tiempo.

**Por ejemplo:** si consideramos el lanzamiento de un dado de seis caras, los eventos "obtener un nmero par" y "obtener un nmero impar" son mutuamente excluyentes, ya que un resultado especfico no puede ser ambos, par e impar al mismo tiempo. En este caso, no hay ningn elemento (resultado del dado) que pertenezca a ambos conjuntos (par e impar), lo que los hace conjuntos disyuntos.

### 2.5. Medida de todo el espacio muestral

---
Medida de todo el espacio muestral

---

Vamos a denotar por $\emptyset$ al conjunto vaco, es decir un conjunto que no tiene elementos.

En nuestro ejemplo tenemos que

$$\mathcal{M}= \text{azul}\cup \text{rojo}\cup \text{gris}$$.

Adems se tiene que

$$
\begin{align}
\text{azul}\cap \text{rojo} &= \emptyset\\
\text{azul}\cap \text{gris} &= \emptyset\\
\text{gris}\cap \text{rojo} &= \emptyset\\
\end{align}
$$

Se dice en esta situacin que los conjuntos son `mutuamente excluyentes`.  De acuerdo con la regla aditiva tenemos que

$$
\text{Prob}[\mathcal{M}] = \text{Prob}[\text{azul}] +  \text{Prob}[\text{rojo}]+  \text{Prob}[\text{gris}] = 5/20 + 7/20 + 8/20 = 1.
$$

Esta es una propiedad general de la probabilidad. El espacio muestral siempre tiene medida de probabilidad 1.

Adems observe que si se tienen eventos disyuntos entre s (mutuamente excluyentes), cuya unin es el espacio muestral, entonces la probabilidad de la unin de todos esos eventos tiene probabilidad 1.



### 2.6. Probabilidad del complemento de un evento

---
Probabilidad del complemento de un evento

---

El complemento de un evento $A$ se denotar por $A^{c}$. Este simplemente el conjunto de elementos del espacio muestral que estn por fuera de $A$. Entonces, es inmediato que $\mathcal{M} = A\cup A^c$. Por lo que

$$
Prob[A^c] = 1 - Prob[A].
$$

Una consecuencia inmediata de esta propiedad es que como $\mathcal{M}^c= \emptyset$, porque el espacio muestral contiene a todos los elementos, entonces $Prob[\emptyset]=0$.


En nuestro ejemplo $impares^c= pares$. Entonces $Prob[\text{impares} ] = 1- 9/20 = 11/20$.

# **<span style="color:blue">3. Probabilidad Condicional</span>**


---
Probabilidad Condicional

---

El concepto de probabilidad condicional es de vital importancia en el estudio del aprendizaje profundo y la inteligencia artificial.

Como el nombre parece indicar, es trata de calcular la probabilidad de un evento sujeto a una restriccin. En realidad es as y la restriccin normalmente est asociada con otro evento.

Para ilustrar el asunto, supongamos que se pregunta por la probabilidad que una bola extrada sea par, dado que la bola es azul.

Se observa entonces, que se da una informacin antes de calcular la probabilidad de ser par. Esta informacin corresponde al evento *azul*. Escribiremos

$$
\text{Prob}(\text{par}|\text{azul})
$$


Para hacer el clculo correcto, se procede de la siguiente manera: Primero se reduce el espacio muestra a *azul*. En el ejemplo se tiene que  

$$
\text{azul} = \{5,7,8,10,16 \}.
$$

Ahora que se ha restringido el espacio muestral a *azul*, se calcula la probabilidad de inters. En este caso *par*. Observe entonces que

$$
\text{Prob}(\text{par}|\text{azul}) = \tfrac{3}{5},
$$

porque en el evento *azul* que tiene 5 elementos hay 3 *pares*.


Puede verificarse que

$$
\text{Prob}(\text{par}|\text{azul}) = \frac{\text{Prob}[\text{par}\cap \text{azul}]}{\text{Prob}[\text{azul}]}
$$

Esta es una regla general, que se enuncia a si. Si $A$ y $B$ son eventos del espacio muestral $\mathcal{M}$, entonces se define $\text{Prob}[A|B]$ como

$$
\text{Prob}[A|B] = \frac{\text{Prob}[A\cap B]}{\text{Prob}[B]}
$$

### 3.1. <span style="color:blue">Regla multiplicativa de la probabilidad</span>

---
Regla multiplicativa de la probabilidad

---

De la definicin de la probabilidad condicional $\text{Prob}[A|B]$ se desprende que

$$
\text{Prob}[A\cap B] = \text{Prob}[B]\times \text{Prob}[A|B]
$$


###### 3.1.1. Ejemplo

---
Ejemplo

---

Con nuestro ejemplo supongamos que se pregunta por la probabilidad de obtener una bola par azul en un experimento.

La solucin es sencilla, por que ya hemos obtenido que $\text{Prob}(\text{par}|\text{azul}) = \tfrac{3}{5}$,
y $\text{Prob}[\text{azul}] = 5/20$. por lo tanto

$$
\text{Prob}[\text{par}\cap\text{azul}] = \text{Prob}[\text{azul}]\times\text{Prob}[\text{par}|\text{azul}] = \tfrac{5}{20}\times \tfrac{3}{5} =  \tfrac{3}{20}
$$

Esto est de acuerdo con la ilustracin de los evento del espacio muestral exhibidos arriba.

### 3.2. <span style="color:blue">Independencia</span>

---
Independencia entre eventos

---

Dos eventos $A$ y $B$ del espacio muestral $\mathcal{M}$ se dicen independientes si


$$
\text{Prob}[A\cap B] = \text{Prob}[A] \times\text{Prob}[B].
$$


Esta definicin es bastante tcnica, pero intuitivamente puede entenderse como que la ocurrencia de un evento no afecta la ocurrencia del otro. Observe que en este caso se tiene que

$$
\text{Prob}[A| B] = \text{Prob}[A].
$$

La siguiente funcin muestra un camino de como calcular la tabla de frecuencias de un conjunto de datos, puestos en una lista de Python.

````python
# Funcin Python para contar los elementos la fecuencia
# de los elementos de una lista, usando un diccionario

def CountFrequency(my_list):

    # Creating an empty dictionary
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
    return freq
````

$$
P(k, n; p) = {n \choose k} p^k (1-p)^{n-k} = \frac{n!}{k!(n-k)!}p^k (1-p)^{n-k}
$$

donde $P(k, n; p)$ representa la probabilidad de obtener $k$ xitos de $n$ intentos con posibilidad **binaria** (por ejemplo, lanzamientos de moneda).

`Ejemplo`: la probabilidad de obtener 4 caras a partir de 10 lanzamientos consecutivos de moneda, est dada por (tomando $p=0.5$, por lo tanto $1-p=0.5$):

$$

$$



Ahora, la probabilidad de obtener $k$ o menos xitos a partir de $n$ intentos est dada por la distribucin acumulada:

$$
C(k, n; p) = \sum_{i=0}^k P(i, n;p) = \sum_{i=0}^k {n \choose i} p^i (1-p)^{n-i}
$$

Por convencin entendemos que:

$$C(k=3,n=6;p=0.5) = P(k \leq 3, n=6, p=0.5 )$$

### 6.1. Distribucin Gausiana

---
Distribuciones Gaussiana

---

Recordemos que la distribucin de probabilidad normal o gaussiana est dada por:

$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

donde:

* $\mu$: media de la distribucin
* $\sigma$: desviacin estandar de la distribucin

`Ejemplo`: considerando una variable aleatoria que sigue una distribucin normal con media $\mu=4$ y desviacin estndar $\sigma=0.3$, la probabilidad de que dicha variable tome el valor de $0.2$ est dada por:

$$P(0.2) =  \frac{1}{0.3 \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{0.2-4}{0.3} \right)^2 \right]}$$

Lo cual en Python se traduce en:

```python
from scipy.stats import norm

norm(mu, sigma).pdf(X)
```

---

As tambin, la distribucion de probabilidad acumulada correspondiente est dada por:

$$C(X) = \int_{x<=X} P(x)dx = \int_{- \infty}^xP(X)dX $$

teniendo en cuenta que $Dom(X) = (-\infty, \infty)$.

`Ejemplo`: considerando una variable aleatoria que sigue una distribucion normal con media $\mu=4$ y desviacin estndar $\sigma=0.3$, la probabilidad de que dicha variable tome el valor de $0.2$ o menos est dada por:

$$C(0.2) = \int_{x<=0.2} P(x)dx = \int_{- \infty}^{0.2}  \left\{\frac{1}{0.3 \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-4}{0.3} \right)^2 \right]} \right\} dX $$

La cual se calcula en Python como:

```python
from scipy.stats import norm

norm(mu, sigma).cdf(X)
```

_**Es importante recordar que la funcin de probabilidad acumulada de la distribucin gaussiana no se puede calcular de forma exacta, ya que la integral anterior no tiene una expresin cerrada conocida. Es decir, los mtodos de integracin conocidos no funcionan para resolver esta integral.**_

Varianza y desviacin estndar de una variable aleatoria

La varianza de una variable aleatoria mide su nivel de predictibilidad. Dicho en otras palabras, que tan dispersos son los valores de la variable aleatoria en relacin con la esperanza matemtica. Denotemos la esperanza de $X$ por $\mu_X$.

$$

### 8.4. Desviacin estndar

---
Desviacin estndar

---

La desviacin estndar de una variable aleatoria se denota $\sigma_X$ y es simplemente la raz cuadrada de la varianza, es decir,

$$
\sigma_X = \sqrt{Var[X]}.
$$

Para entender el porque la varianza mide el nivel de predictibilidad de una variable aleatoria consideremos una variable Bernoulli $X$. Escribimos  $X \sim \text{Ber}(\pi)$. En este caso puede verificarse que la varianza es dada por

$$
Var(X) = \pi(1-\pi).
$$

El siguiente grfico muestra la varianza de las distribuciones  Bernoulli con valorea distintos del parmetro $\pi$.

````python
import numpy as np
import matplotlib.pyplot as plt

pi = np.linspace(0,1,100)
var = pi*(1-pi)
plt.plot(pi,var)
plt.title('Funcion de varianza de distribciones Bernoulli')
plt.xlabel( '$\pi$')
plt.ylabel('Var')
plt.show()
````

Se observa entonces que la mxima varianza se alcanza para el caso $\pi=0.5$. Es porque en este caso, es mas difcil predecir el resultado del experimento, dado que la probabilidad de *acierto* y *fallo* son la misma. Pero a medida que el valor $\pi$ est cerca de  cero o uno, la varianza desciende hasta cero. Por ejemplo si $\pi = 0.9$, casi siempre se obtendr *xito*. Esto significa que la variable es ms predictible para valores muy altos o muy bajos de $\pi$. **A menor varianza mayor precisin para predecir el resultado del experimento y viceversa**.

# **<span style="color:blue">9. Entropa</span>**

### <span style="color:blue"> 9.1. Entropa de una variable aleatoria</span>

---
Entropa de una variable aleatoria

---

Similar a la varianza, la entropa de una variable aleatoria mide su grado de predictibilidad desde el punto de vista de la teora de informacin de Shanon. Dada una variable aleatoria discreta $X=\{x_1,x_2,\ldots \}$ con funcin de probabilidad $p_X =\{p_1, p_2,\ldots \}$, la entropa de $X$ se define por

$$
H(X) = -\sum_i p_i\log p_i = -\mathbb{E}[\log P_X]
$$

Es usual utilizar la base 2 o la base de los logaritmos Neperianos. En el primer caso, la unidad de medida de la entropa de denomina bit.  En el caso de los logaritmos Neperianos, la unidad de medida se acostumbra a llamar nat. En realidad pasar de una base a otra es simplemente un cambio de escala. Estructuralmente miden lo mismo en distintas unidades.

Como en el caso de la varianza ilustramos el concepto para la distribucin Bernoulli.

````python
import numpy as np
import matplotlib.pyplot as plt

pi = np.linspace(0.0000001,0.999999,100)
H = -(pi*np.log(pi) + (1-pi)*np.log(1-pi))
plt.plot(pi,H)
plt.title('Entropa de distribuciones Bernoulli')
plt.xlabel( '$\pi$')
plt.ylabel('Entropa')
plt.show()
````

Como se observa la entropa mide de la misma forma que la varianza pero en una escala diferente. Por otro lado observe que en el clculo de la entropa no se utilizan los valores que toma la variable aleatoria. Solamente se requiere la funcin de probabilidad de la variable.

En palabras se dice que **la entropa no depende de la escala de la  variable**. Solamente de su estructura de probabilidad. Por su parte la varianza si es dependiente de la escala de la variable aleatoria.

Aunque el contenido de informacin proporciona informacin interesante sobre eventos especficos $ x $, en algunos casos nos gustara conocer el contenido de informacin de una distribucin de probabilidad. En este asunto, lo ms razonable sera estimar la informacin esperada. Esta estimacin se conoce como entropa.

Supongamos que $X$ es una variable aleatoria discreta. Le entropa (de Shanon) de la variable aleatoria $X$, o lo que es lo mismo, la entropa de la distribucin asociada a $X$ se define por

$$H(X)=\sum_{x\in \Omega} P(X=x)I(x)=-\sum_{x\in \Omega}P(X=x)\log{P(X=x)} = -\sum_{i} p_i\log p_i.$$




1. En ocasiones, la entropa de Shanon para una variable aleatoria $ X $ con probabilidades $ p_1, \ldots, p_M $ se denota $ H (p_1, \ldots, p_M) $. Por ejemplo, la entropa de una distribucin de Bernoulli a veces se denota $H(p,q)= H(p,1-p)$.
2. Algunos autores llaman a  $H(X)$ como *incertidumbre*.
3. $H(X)$ Es una esperanza. La entropa es solo la **sorpresa media** de la variable aleatoria $H(X)$.
